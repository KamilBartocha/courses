## 1. 🔹 List comprehensions

**List comprehension** to zwięzły sposób tworzenia listy na podstawie
iterowalnego obiektu - w jednej linii, z opcjonalnym filtrem.

```python
new_list = [wyrażenie for element in iterable if warunek]
```

| | Podejście imperatywne | List comprehension |
|---|---|---|
| Tworzenie | pętla `for` + `append()` | `[x for x in ...]` |
| Filtrowanie | `if` wewnątrz pętli | `if` na końcu wyrażenia |
| Czytelność | wieloliniowe | jednoliniowe |
| Wydajność | wolniejsze | szybsze (wbudowana optymalizacja) |

> 💡 List comprehension zwraca **nową listę** - nie modyfikuje
> oryginalnej kolekcji.

````python
# Podejście imperatywne
squares_imperative = []
for x in range(10):
    squares_imperative.append(x ** 2)
print(squares_imperative)

# List comprehension - to samo, krócej
squares = [x ** 2 for x in range(10)]
print(squares)
````

````python
# Z filtrem - tylko parzyste
even_triples = [x * 3 for x in range(10) if x % 2 == 0]
print(even_triples)

# Transformacja napisów
words = ["hello", "world", "python"]
upper_words = [word.upper() for word in words]
print(upper_words)

# Zagnieżdżone - spłaszczenie macierzy 2D
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [num for row in matrix for num in row]
print(flat)
````

## 2. 🔹 Dictionary comprehensions

**Dictionary comprehension** tworzy nowy słownik (dict) w jednej linii.

```python
new_dict = {klucz: wartość for element in iterable if warunek}
```

Przydatne wzorce:
- `{k: v for k, v in dict.items()}` - iteracja po parach klucz-wartość
- `{k: v for k, v in dict.items() if v > 0}` - z filtrem
- `{v: k for k, v in dict.items()}` - odwrócenie słownika

> 💡 Do iteracji po istniejącym słowniku zawsze używamy `.items()`,
> które zwraca pary `(klucz, wartość)`.

````python
# Kwadraty liczb jako słownik
squares_dict = {x: x ** 2 for x in range(1, 6)}
print(squares_dict)

# Długości słów, ale tylko zaczynających się na 'a'
words = ['apple', 'banana', 'cherry', 'ananas']
length_dict = {word: len(word) for word in words if word.startswith('a')}
print(length_dict)

# Odwrócenie słownika - .items() explicite
original = {'a': 1, 'b': 2, 'c': 3}
inverted = {v: k for k, v in original.items()}
print(inverted)
````

## 3. 🔹 Set i Generator comprehensions

### Set comprehension

Tworzy **zbiór (set)** - kolekcję unikalnych elementów:

```python
new_set = {wyrażenie for element in iterable if warunek}
```

### Generator comprehension

Tworzy **generator** - obiekt generujący elementy na żądanie,
bez przechowywania wszystkich w pamięci:

```python
generator = (wyrażenie for element in iterable if warunek)
```

| | List `[...]` | Generator `(...)` |
|---|---|---|
| Pamięć | przechowuje wszystkie | generuje po jednym |
| Iteracja | wielokrotna | jednorazowa |
| Dostęp | po indeksie `[i]` | tylko `next()` lub pętla |
| Kiedy używać | potrzebujemy listy | duże dane, jedno przejście |

> 💡 Funkcja ze słowem `yield` jest generatorem - zawiesza swój
> stan po każdym `yield` i wznawia przy następnym `next()`.

````python
# Set comprehension - unikalne kwadraty
squares_set = {x ** 2 for x in range(10)}
print(squares_set)  # kolejność nieustalona - to zbiór!

# Set usuwa duplikaty
words = ['apple', 'banana', 'apple', 'cherry', 'banana']
unique_first_chars = {word[0] for word in words}
print(unique_first_chars)
````

````python
# Generator comprehension vs list comprehension
squares_list = [x**2 for x in range(1000)]   # tworzy całą listę
squares_gen  = (x**2 for x in range(1000))   # nie tworzy nic jeszcze

print(type(squares_list))   # <class 'list'>
print(type(squares_gen))    # <class 'generator'>

# Generator - pobieramy elementy jeden po jednym
gen = (x**2 for x in range(5))
print(next(gen))   # 0
print(next(gen))   # 1
print(list(gen))   # [4, 9, 16] - pozostałe elementy
````

````python
# Funkcja generatora z yield
def count_up_to(max_value):
    count = 1
    while count <= max_value:
        yield count   # zawiesza funkcję, oddaje wartość
        count += 1

gen = count_up_to(5)
print(list(gen))   # [1, 2, 3, 4, 5]

# next() - pobiera kolejne wartości
gen2 = count_up_to(3)
print(next(gen2))  # 1
print(next(gen2))  # 2
print(next(gen2))  # 3
# print(next(gen2))  # StopIteration - wyczerpany generator
````

## 4. 🔹 Funkcje lambda

**Funkcja lambda** to anonimowa (nienazwana) funkcja jednolinijkowa.
Tworzymy ją słowem kluczowym `lambda`:

```python
lambda argumenty: wyrażenie
```

| | `def` | `lambda` |
|---|---|---|
| Nazwa | wymagana | brak (anonimowa) |
| Ciało | wiele linii | jedno wyrażenie |
| `return` | explicite | niejawne (zwraca wyrażenie) |
| Kiedy | złożona logika | proste one-linery |

Lambda najczęściej stosujemy jako argument `key=` w `sorted()`,
`min()`, `max()` - wszędzie tam, gdzie potrzebujemy krótkiej
jednorazowej funkcji.

> 💡 Lambdy nie mogą zawierać instrukcji (`if/else` jako wyrażenie
> tak, ale nie jako instrukcja - bez `elif`). Dla złożonej logiki
> używamy zwykłej funkcji `def`.

````python
# Podstawowe lambdy
add_ten  = lambda x: x + 10
multiply = lambda x, y: x * y
max_val  = lambda x, y: x if x > y else y

print(add_ten(5))       # 15
print(multiply(4, 5))   # 20
print(max_val(7, 10))   # 10
````

````python
# Lambda jako key= w sorted()
tuples = [(1, 3), (2, 1), (4, 2)]
sorted_by_second = sorted(tuples, key=lambda t: t[1])
print(sorted_by_second)   # [(2, 1), (4, 2), (1, 3)]

# Sortowanie listy słowników
people = [
    {"name": "Alice", "age": 30},
    {"name": "Bob",   "age": 25},
    {"name": "Carol", "age": 35},
]
sorted_people = sorted(people, key=lambda p: p["age"])
for p in sorted_people:
    print(p["name"], p["age"])
````

## 5. 🔹 `map()` i `filter()`

Obie funkcje są narzędziami **programowania funkcyjnego** - przyjmują
funkcję i iterable, zwracają **lazy iterator** (nie listę).

```python
map(funkcja, iterable)       # stosuje funkcję do każdego elementu
filter(funkcja, iterable)    # zatrzymuje elementy, dla których f() = True
```

| | `map()` | `filter()` |
|---|---|---|
| Cel | transformacja | filtrowanie |
| Wynik | tyle samo elementów | <= tyle samo elementów |
| Funkcja zwraca | dowolna wartość | `True`/`False` |
| Alternatywa | list comp `[f(x) for x in ...]` | list comp z `if` |

> 💡 `map()` i `filter()` zwracają iteratory - stosujemy `list()`
> aby uzyskać listę. Można je łączyć: `map(f, filter(g, data))`.

````python
# map() - stosuje funkcję do każdego elementu
numbers = [1, 2, 3, 4, 5]

# Z nazwaną funkcją
def square(x):
    return x ** 2

squared = list(map(square, numbers))
print(squared)   # [1, 4, 9, 16, 25]

# Z lambdą - częściej spotykane
squared = list(map(lambda x: x ** 2, numbers))
print(squared)   # [1, 4, 9, 16, 25]

# map() na dwóch listach jednocześnie
a = [1, 2, 3]
b = [4, 5, 6]
sums = list(map(lambda x, y: x + y, a, b))
print(sums)      # [5, 7, 9]
````

````python
# filter() - zachowuje elementy spełniające warunek
numbers = [1, 2, 3, 4, 5, 6]

# Z nazwaną funkcją
def is_even(x):
    return x % 2 == 0

evens = list(filter(is_even, numbers))
print(evens)   # [2, 4, 6]

# Z lambdą
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)   # [2, 4, 6]

# Filtrowanie napisów
words = ["cat", "dog", "elephant", "fish"]
long_words = list(filter(lambda s: len(s) > 3, words))
print(long_words)   # ['elephant', 'fish']
````

````python
# map() lazy evaluation - ważna pułapka!
numbers = [1, 2, 3, 4, 5]
squared_map = map(lambda x: x ** 2, numbers)

numbers[0] = 00   # zmieniamy oryginalną listę

print(list(squared_map))   # [10000, 4, 9, 16, 25] - lazy!
# map() przetwarza dopiero przy iteracji, nie przy tworzeniu
````
