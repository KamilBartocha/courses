## 1. 🔹 Format JSON i typy danych

JSON (JavaScript Object Notation) to tekstowy format wymiany danych.
Używamy go wszędzie tam, gdzie trzeba przesłać lub zapisać dane w sposób
czytelny dla człowieka i łatwy do przetworzenia przez program.

Trzy zasady formatu JSON:
- klucze zawsze w podwójnych cudzysłowach (`"key"`)
- brak komentarzy i brak przecinka po ostatnim elemencie
- wartości logiczne pisane małą literą (`true`, `false`, `null`)

### Mapowanie typów Python ↔ JSON

| Python | JSON |
|--------|------|
| `dict` | obiekt `{}` |
| `list`, `tuple` | tablica `[]` |
| `str` | string `"..."` |
| `int`, `float` | number |
| `True` / `False` | `true` / `false` |
| `None` | `null` |

```python
import json

# Python dict -> JSON string
person = {
    "name": "Alice",
    "age": 30,
    "active": True,
    "score": None
}

json_str = json.dumps(person, indent=2)
print(json_str)
# Uwaga: True -> true, None -> null

# JSON string -> Python dict
restored = json.loads(json_str)
print(type(restored["active"]))   # <class 'bool'>
print(restored["score"] is None)  # True
```

## 2. 🔹 Serializacja i deserializacja

Moduł `json` udostępnia cztery podstawowe funkcje. Trick do zapamiętania:
litera **`s`** na końcu nazwy oznacza **string** (operacja na napisie),
brak `s` oznacza **file** (operacja na pliku).

| Funkcja | Kierunek | Gdzie |
|---------|----------|-------|
| `json.dumps(obj)` | Python -> JSON | napis (string) |
| `json.loads(s)` | JSON -> Python | napis (string) |
| `json.dump(obj, f)` | Python -> JSON | plik |
| `json.load(f)` | JSON -> Python | plik |

> 💡 **Tip:** `json.dumps()` i `json.loads()` to najczęściej używana
> para - np. przy pracy z odpowiedziami API lub testowaniu bez pliku.

```python
import json

# --- Praca na napisach ---
config = {"host": "localhost", "port": 8080, "debug": False}

# Serializacja: Python dict -> JSON string
json_str = json.dumps(config, indent=2)
print(json_str)

# Deserializacja: JSON string -> Python dict
loaded = json.loads(json_str)
print(loaded["port"])   # 8080
print(type(loaded["debug"]))   # <class 'bool'>
```

```python
import json

# --- Praca na plikach ---
data = {"name": "Alice", "scores": [95, 87, 92]}

# Zapis do pliku (serializacja)
with open("sample.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)

# Odczyt z pliku (deserializacja)
with open("sample.json", "r", encoding="utf-8") as f:
    loaded = json.load(f)

print(loaded["name"])     # Alice
print(loaded["scores"])   # [95, 87, 92]
```

## 3. 🔹 Praca z plikami JSON

Przy zapisie do pliku warto pamiętać o kilku parametrach:

| Parametr | Wartość domyślna | Kiedy zmienić |
|----------|-----------------|---------------|
| `indent` | `None` (brak wcięć) | gdy plik ma być czytelny |
| `ensure_ascii` | `True` | ustaw `False` dla polskich znaków |
| `sort_keys` | `False` | gdy chcemy stały porządek kluczy |

Zawsze otwieramy plik z `encoding='utf-8'`, żeby obsługiwać
polskie znaki i inne znaki spoza ASCII.

> 💡 **Tip:** Tryb `'w'` nadpisuje plik, `'a'` dopisuje na końcu.
> Przy JSON prawie zawsze chcemy `'w'`, żeby plik był poprawnym JSON-em.

```python
import json

# Typowy wzorzec zapisu konfiguracji
config = {
    "database": {
        "host": "localhost",
        "port": 5432,
        "name": "appdb"
    },
    "server": {
        "host": "0.0.0.0",
        "port": 8080
    }
}

with open("config.json", "w", encoding="utf-8") as f:
    json.dump(config, f, indent=2, ensure_ascii=False)

# Typowy wzorzec odczytu
with open("config.json", encoding="utf-8") as f:
    loaded = json.load(f)

print(loaded["database"]["port"])   # 5432
print(loaded["server"]["host"])     # 0.0.0.0
```

## 4. 🔹 Obsługa błędów

Dwa najczęstsze wyjątki przy pracy z JSON:

| Wyjątek | Przyczyna |
|---------|----------|
| `json.JSONDecodeError` | niepoprawny format JSON (brak `"`, przecinek) |
| `FileNotFoundError` | plik nie istnieje |
| `KeyError` | klucz nie istnieje w załadowanym słowniku |

Dobrą praktyką jest napisanie pomocniczej funkcji `safe_load`, która
zwraca `None` zamiast rzucać wyjątek - szczególnie przy wczytywaniu
zewnętrznych plików konfiguracyjnych.

> 💡 **Tip:** `json.JSONDecodeError` jest podklasą `ValueError`.
> Możemy łapać `ValueError`, ale `JSONDecodeError` daje dokładniejszy
> komunikat z numerem linii błędu.

```python
import json

# JSONDecodeError - brakujacy cudzyslow
broken = '{"name": "Alice", city: "Warsaw"}'

try:
    data = json.loads(broken)
except json.JSONDecodeError as e:
    print(f"JSON error: {e}")

# FileNotFoundError
try:
    with open("nonexistent.json") as f:
        data = json.load(f)
except FileNotFoundError as e:
    print(f"File error: {e}")

# Bezpieczna funkcja pomocnicza
def safe_load_json(path):
    try:
        with open(path, encoding="utf-8") as f:
            return json.load(f)
    except (FileNotFoundError, json.JSONDecodeError) as e:
        print(f"Warning: {e}")
        return None

result = safe_load_json("nonexistent.json")
print(result)   # None
```
