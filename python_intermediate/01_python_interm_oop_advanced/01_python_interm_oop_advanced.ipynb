## 1. 🔹 Podstawy OOP

Programowanie obiektowe (Object-Oriented Programming, OOP) to
paradygmat, w którym organizujemy kod wokół **obiektów** - instancji
klas. **Klasa (class)** to szablon definiujący atrybuty i metody.
**Obiekt (object)** to konkretny egzemplarz klasy.

Kluczowe pojęcia:

- **klasa (class)** - szablon opisujący strukturę i zachowanie obiektów
- **obiekt (object)** - instancja klasy, konkretny egzemplarz
- **atrybut (attribute)** - zmienna przechowywana w obiekcie
- **metoda (method)** - funkcja zdefiniowana wewnątrz klasy
- **konstruktor (constructor)** - metoda `__init__`, wywoływana
  automatycznie przy tworzeniu nowego obiektu

> 💡 Każda metoda instancji przyjmuje `self` jako pierwszy parametr -
> jest to odwołanie do konkretnego obiektu, na którym wywołujemy metodę.

```python
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def description(self):
        return f"{self.make} {self.model}, year: {self.year}"


car1 = Car("BMW", "330i", 2020)
print(car1.description())
print(car1.make)
```

```python
class Dog:
    def __init__(self, name, breed, age):
        self.name = name
        self.breed = breed
        self.age = age

    def bark(self):
        return f"{self.name} says Woof!"


dog1 = Dog("Buddy", "Golden Retriever", 3)
print(dog1.bark())
```

```python
class Student:
    def __init__(self, name, student_id, grades):
        self.name = name
        self.student_id = student_id
        self.grades = grades

    def average_grade(self):
        return sum(self.grades) / len(self.grades)


student1 = Student("Alice", "S12345", [90, 85, 88, 92])
print(student1.average_grade())
```

## 2. 🔹 Dziedziczenie i `super()`

Dziedziczenie (inheritance) pozwala tworzyć nową klasę na podstawie
istniejącej. Klasa potomna dziedziczy atrybuty i metody klasy bazowej
i może je rozszerzać lub **nadpisywać (override)**.

**Polimorfizm (polymorphism)** to zdolność różnych klas do implementowania
tej samej metody w odmienny sposób.

Funkcja `super()` pozwala odwołać się do klasy nadrzędnej:
- `super().__init__(...)` - wywołujemy konstruktor klasy bazowej
- `super().method()` - wywołujemy i rozszerzamy metodę klasy bazowej

**MRO (Method Resolution Order)** to kolejność, w jakiej Python
przeszukuje klasy przy wywołaniu metody. Sprawdzimy ją przez
`ClassName.__mro__`.

| | Bez `super()` | Z `super()` |
|---|---|---|
| Konstruktor | Powtarzamy kod inicjalizacji | `super().__init__(...)` |
| Metoda | Nie korzystamy z logiki rodzica | Rozszerzamy logikę rodzica |
| Wielokrotne dziedz. | Problematyczne | Obsługuje MRO poprawnie |

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def description(self):
        return f"{self.name}, age: {self.age}"

    def sound(self):
        raise NotImplementedError("Subclass must implement sound()")


class Dog(Animal):
    def sound(self):
        return "Woof!"


class Cat(Animal):
    def sound(self):
        return "Meow!"


dog = Dog("Rex", 5)
cat = Cat("Whiskers", 3)
print(dog.description())
print(dog.sound())
print(cat.description())
print(cat.sound())
```

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def description(self):
        return f"{self.make} {self.model}, year: {self.year}"


class Car(Vehicle):
    def __init__(self, make, model, year, num_doors=4):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def description(self):
        return f"{super().description()} - {self.num_doors} doors"


class Motorcycle(Vehicle):
    def __init__(self, make, model, year, style=None):
        super().__init__(make, model, year)
        self.style = style

    def description(self):
        return f"{super().description()} - style: {self.style}"


car = Car("Toyota", "Corolla", 2020, num_doors=4)
moto = Motorcycle("Yamaha", "MT-07", 2019, "Naked")
print(car.description())
print(moto.description())
```

```python
class A:
    def __init__(self):
        print("Initializing A")

    def method_a(self):
        print("Method from A")


class B(A):
    def __init__(self):
        super().__init__()
        print("Initializing B")

    def method_b(self):
        print("Method from B")


class C(B):
    def __init__(self):
        super().__init__()
        print("Initializing C")

    def method_c(self):
        print("Method from C")


obj = C()
obj.method_a()
obj.method_b()
obj.method_c()
print(C.__mro__)
```

## 3. 🔹 Hermetyzacja (enkapsulacja)

Hermetyzacja (Encapsulation) polega na ukrywaniu wewnętrznej
implementacji klasy i udostępnianiu tylko niezbędnych interfejsów.

W Pythonie oznaczamy poziom dostępu przez prefiks nazwy atrybutu:

| Prefiks | Przykład | Znaczenie |
|---|---|---|
| brak | `self.name` | publiczny - dostępny wszędzie |
| `_` | `self._name` | chroniony - konwencja, zalecamy nie edytować |
| `__` | `self.__name` | prywatny - name mangling, blokuje dostęp |

Python stosuje **name mangling**: `__attr` staje się `_ClassName__attr`.
Dostęp przez `obj._ClassName__attr` jest technicznie możliwy, ale
łamie zasadę hermetyzacji.

> 💡 Hermetyzacja chroni dane przed przypadkową modyfikacją z zewnątrz
> klasy, co zwiększa bezpieczeństwo i integralność obiektów.

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance  # private

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance: {self.__balance}")
        else:
            print("Amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount}. New balance: {self.__balance}")
        else:
            print("Insufficient funds or invalid amount.")

    def check_balance(self):
        return f"Balance: {self.__balance}"


account = BankAccount("Alice", 1000)
account.deposit(500)
account.withdraw(300)
print(account.check_balance())

# Direct access causes AttributeError:
# print(account.__balance)

# Name mangling - not recommended:
print(account._BankAccount__balance)
```

```python
class Rectangle:
    def __init__(self, width, height):
        self.__width = width
        self.__height = height

    def set_dimensions(self, width, height):
        if width > 0 and height > 0:
            self.__width = width
            self.__height = height
        else:
            print("Width and height must be positive.")

    def area(self):
        return self.__width * self.__height

    def perimeter(self):
        return 2 * (self.__width + self.__height)

    def get_dimensions(self):
        return (self.__width, self.__height)


rect = Rectangle(5, 10)
print(f"Area: {rect.area()}")
print(f"Perimeter: {rect.perimeter()}")
print(f"Dimensions: {rect.get_dimensions()}")
rect.set_dimensions(3, 4)
print(f"New area: {rect.area()}")
```

## 4. 🔹 Metody specjalne (dunder)

Metody dunder (double underscore) zaczynają się i kończą `__`.
Definiują zachowanie obiektów w różnych kontekstach Pythona -
umożliwiają **przeciążanie operatorów (operator overloading)**.

| Metoda | Wywołanie | Opis |
|---|---|---|
| `__init__` | `obj = Class()` | konstruktor obiektu |
| `__str__` | `str(obj)`, `print(obj)` | reprezentacja dla użytkownika |
| `__repr__` | `repr(obj)` | reprezentacja dla dewelopera |
| `__add__` | `obj1 + obj2` | operator dodawania `+` |
| `__eq__` | `obj1 == obj2` | operator równości `==` |
| `__len__` | `len(obj)` | długość obiektu |

> 💡 `__str__` jest czytelne dla użytkownika, `__repr__` precyzyjne
> dla dewelopera. Jeśli zdefiniujemy tylko `__repr__`, Python użyje
> go też dla `str()`. Definiujemy oba, gdy potrzebujemy różnych wersji.

```python
class PersonWithStr:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Person(Name: {self.name}, Age: {self.age})"

    def __repr__(self):
        return f"Person(name={self.name!r}, age={self.age})"


class PersonNoStr:
    def __init__(self, name, age):
        self.name = name
        self.age = age


p1 = PersonWithStr("Alice", 30)
p2 = PersonNoStr("Bob", 27)
print(p1)          # Person(Name: Alice, Age: 30)
print(repr(p1))    # Person(name='Alice', age=30)
print(p2)          # <__main__.PersonNoStr object at 0x...>
```

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f"Vector(x={self.x}, y={self.y})"

    def __str__(self):
        return f"({self.x}, {self.y})"

    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        return NotImplemented

    def __eq__(self, other):
        if isinstance(other, Vector):
            return self.x == other.x and self.y == other.y
        return False


v1 = Vector(2, 3)
v2 = Vector(4, 5)
print(v1)                   # (2, 3)
print(repr(v1))             # Vector(x=2, y=3)
print(v1 + v2)              # (6, 8)
print(v1 == Vector(2, 3))   # True
print(v1 == v2)             # False
```

```python
class Matrix:
    def __init__(self, rows):
        self.rows = rows
        self.num_rows = len(rows)
        self.num_cols = len(rows[0]) if rows else 0

    def __repr__(self):
        return f"Matrix(rows={self.rows})"

    def __str__(self):
        return '\n'.join(['\t'.join(map(str, row)) for row in self.rows])

    def __add__(self, other):
        if isinstance(other, Matrix) and (
            self.num_rows == other.num_rows
            and self.num_cols == other.num_cols
        ):
            result = [
                [self.rows[i][j] + other.rows[i][j]
                 for j in range(self.num_cols)]
                for i in range(self.num_rows)
            ]
            return Matrix(result)
        return NotImplemented

    def __mul__(self, other):
        if isinstance(other, (int, float)):
            result = [[el * other for el in row] for row in self.rows]
            return Matrix(result)
        if isinstance(other, Matrix) and self.num_cols == other.num_rows:
            result = [
                [sum(self.rows[i][k] * other.rows[k][j]
                     for k in range(self.num_cols))
                 for j in range(other.num_cols)]
                for i in range(self.num_rows)
            ]
            return Matrix(result)
        return NotImplemented


m1 = Matrix([[1, 2], [3, 4]])
m2 = Matrix([[5, 6], [7, 8]])
print("m1 + m2:")
print(m1 + m2)
print("m1 * 2:")
print(m1 * 2)
print("m1 * m2:")
print(m1 * m2)
```

## 5. 🔹 Metody klasowe: `self` vs `cls`

W Pythonie wyróżniamy trzy rodzaje metod:

| Rodzaj | Dekorator | Pierwszy param | Dostęp do |
|---|---|---|---|
| Metoda instancji | brak | `self` | instancji i klasy |
| Metoda klasowa | `@classmethod` | `cls` | tylko klasy |
| Metoda statyczna | `@staticmethod` | brak | brak |

- **`self`** - odwołanie do konkretnego obiektu (instancji)
- **`cls`** - odwołanie do samej klasy (nie do instancji)

Metody klasowe (`@classmethod`) stosujemy gdy:
- operujemy na atrybutach wspólnych dla wszystkich instancji
- tworzymy **alternatywne konstruktory** (factory methods)

> 💡 Atrybuty klasy (class attributes) są wspólne dla wszystkich
> instancji - zmiana przez jedną instancję widoczna jest dla wszystkich.

```python
class Car:
    def __init__(self, make, model):
        self.make = make    # attribute of this specific instance
        self.model = model

    def description(self):
        # self refers to the specific car object
        return f"{self.make} {self.model}"


car1 = Car("Toyota", "Corolla")
car2 = Car("Honda", "Civic")
print(car1.description())  # Toyota Corolla
print(car2.description())  # Honda Civic
```

```python
class Car:
    car_count = 0  # class attribute - shared by all instances

    def __init__(self, make, model):
        self.make = make
        self.model = model
        Car.car_count += 1

    @classmethod
    def get_count(cls):
        return cls.car_count


car1 = Car("Toyota", "Corolla")
car2 = Car("Honda", "Civic")
print(Car.get_count())   # 2
car3 = Car("BMW", "3 Series")
print(Car.get_count())   # 3
```

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    @classmethod
    def from_fahrenheit(cls, fahrenheit):
        celsius = (fahrenheit - 32) * 5 / 9
        return cls(celsius)

    @classmethod
    def from_kelvin(cls, kelvin):
        celsius = kelvin - 273.15
        return cls(celsius)

    def __str__(self):
        return f"{self.celsius:.2f}°C"


t1 = Temperature(00)
t2 = Temperature.from_fahrenheit(212)
t3 = Temperature.from_kelvin(373.15)
print(t1)   # 00.00°C
print(t2)   # 00.00°C
print(t3)   # 00.00°C
```
