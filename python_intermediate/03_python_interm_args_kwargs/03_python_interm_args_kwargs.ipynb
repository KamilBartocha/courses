## 1. 🔹 `*args`

`*args` to konwencja przyjmowania **zmiennej liczby argumentów
pozycyjnych**. Wewnątrz funkcji `args` jest dostępne jako **krotka
(tuple)**.

```python
def funkcja(*args):
    # args to krotka wszystkich przekazanych wartości
    for arg in args:
        print(arg)
```

| Aspekt | Szczegół |
|---|---|
| Typ wewnątrz funkcji | `tuple` |
| Kolejność w sygnaturze | po zwykłych parametrach |
| Nazwa | konwencja `args`, może być dowolna |
| Dostęp | iteracja, indeksowanie, `len()` |

> 💡 Gwiazda `*` to operator - `args` to tylko konwencja nazewnicza.
> Możemy napisać `*numbers`, `*values` - ważna jest tylko gwiazdka.

````python
def print_args(*args):
    print(args)
    print(type(args))   # <class 'tuple'>
    for arg in args:
        print(arg)

print_args(1, 2, 3, "hello", [1, 2, 3])
````

````python
def add_all(*args):
    total = 0
    for value in args:
        total += value
    return total

print(add_all(1, 1, 1, 1))       # 4
print(add_all(1, 1, 2, 3, 5, 6)) # 18

# Krócej - z sum()
def add_all_v2(*args):
    return sum(args)

print(add_all_v2(10, 20, 30))     # 60
````

````python
def concat_strings(*args):
    return " ".join(args)

print(concat_strings("Hello", "world!"))      # Hello world!
print(concat_strings("Python", "is", "fun"))  # Python is fun
````

## 2. 🔹 `**kwargs`

`**kwargs` to konwencja przyjmowania **zmiennej liczby argumentów
słownikowych** (pary klucz-wartość). Wewnątrz funkcji `kwargs`
jest dostępne jako **słownik (dict)**.

```python
def funkcja(**kwargs):
    # kwargs to dict wszystkich przekazanych par klucz=wartość
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

| Aspekt | Szczegół |
|---|---|
| Typ wewnątrz funkcji | `dict` |
| Kolejność w sygnaturze | po `*args` |
| Nazwa | konwencja `kwargs`, może być dowolna |
| Dostęp | `.items()`, `.keys()`, `.values()`, `kwargs[key]` |

> 💡 Do iteracji po `kwargs` używamy `.items()`, które zwraca
> pary `(klucz, wartość)` - tak samo jak w zwykłym słowniku.

````python
def print_kwargs(**kwargs):
    print(kwargs)
    print(type(kwargs))   # <class 'dict'>
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_kwargs(name='Alice', age=30, city='New York')
````

````python
def configure_app(**settings):
    defaults = {
        'debug': True,
        'log_level': 'INFO',
        'host': 'localhost',
        'port': 8080,
    }
    defaults.update(settings)   # nadpisujemy wartości domyślne
    return defaults

print(configure_app(debug=False, port=5000))
print(configure_app())
````

## 3. 🔹 Łączenie `*args` i `**kwargs`

Możemy łączyć zwykłe parametry, `*args` i `**kwargs` w jednej
funkcji. Ważna jest **kolejność** w sygnaturze:

```python
def f(param1, param2, *args, kwonly=None, **kwargs):
    pass
```

| Kolejność | Typ parametru | Przykład |
|---|---|---|
| 1 | wymagane pozycyjne | `a, b` |
| 2 | `*args` | `*args` |
| 3 | keyword-only (po `*args`) | `debug=False` |
| 4 | `**kwargs` | `**kwargs` |

> 💡 Parametry po `*args` stają się **keyword-only** - można je
> przekazać wyłącznie przez nazwę: `f(1, 2, debug=True)`.

````python
def example(arg1, arg2, *args, kwarg1=None, **kwargs):
    print(f"arg1={arg1}, arg2={arg2}")
    print(f"args={args}")
    print(f"kwarg1={kwarg1}")
    print(f"kwargs={kwargs}")

example(1, 2, 3, 4, 5,
        kwarg1='value1',
        extra1='extra1', extra2='extra2')
````

````python
def report(title, *args, **kwargs):
    print(f"Title: {title}")
    print("Arguments:", args)
    print("Keyword Arguments:", kwargs)

report('Report 1', 10, 20, 30,
       status='completed', author='John')
````

## 4. 🔹 Rozpakowywanie `*` i `**`

Operator `*` i `**` działają również **poza definicją funkcji** -
do rozpakowywania (unpacking) kolekcji.

| Użycie | Opis | Przykład |
|---|---|---|
| `*list` w wywołaniu | rozpakuj listę jako args | `f(*[1, 2, 3])` |
| `**dict` w wywołaniu | rozpakuj dict jako kwargs | `f(**{'a': 1})` |
| `[*a, *b]` | połącz listy | `[*[1,2], *[3,4]]` |
| `{**a, **b}` | połącz/nadpisz dykty | `{**d1, **d2}` |
| `a, *b, c = [...]` | przypisanie z resztą | `first, *rest, last` |

> 💡 Przy `{**d1, **d2}` klucze z `d2` **nadpisują** klucze z `d1`.
> Kolejność ma znaczenie!

````python
# Łączenie słowników przez **
defaults = {'debug': True, 'host': 'localhost', 'port': 8080}
overrides = {'debug': False, 'port': 5000}

config = {**defaults, **overrides}   # overrides wygrywa
print(config)
# {'debug': False, 'host': 'localhost', 'port': 5000}

config2 = {**overrides, **defaults}  # defaults wygrywa
print(config2)
# {'debug': True, 'port': 8080, 'host': 'localhost'}
````

````python
# Łączenie list przez *
fruits = ["apple", "banana"]
veggies = ["carrot", "broccoli"]

all_foods = [*fruits, *veggies]
print(all_foods)

# Wstawianie w środek
extras = ["cherry", *fruits, "date"]
print(extras)
````

````python
# Przypisanie z * - extended unpacking
numbers = [1, 2, 3, 4, 5]

first, *rest = numbers
print(first)   # 1
print(rest)    # [2, 3, 4, 5]

first, *middle, last = numbers
print(first)    # 1
print(middle)   # [2, 3, 4]
print(last)     # 5
````

````python
# Rozpakowywanie przy wywołaniu funkcji
def add(a, b, c):
    return a + b + c

values = [1, 2, 3]
print(add(*values))     # 6  - odpowiednik add(1, 2, 3)

params = {'a': 10, 'b': 20, 'c': 30}
print(add(**params))    # 60 - odpowiednik add(a=10, b=20, c=30)
````
