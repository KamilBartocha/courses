## 1. 🔹 try / except / else / finally

Wyjątek (exception) to obiekt sygnalizujący błąd lub nieoczekiwaną
sytuację podczas działania programu. Zamiast przerywać program,
możemy wyjątek *przechwycić* i obsłużyć w kontrolowany sposób.

Pełna struktura bloku:

```
try:
    # kod, który może zgłosić wyjątek
except SomeError as e:
    # wykonywany gdy wyjątek wystąpił
else:
    # wykonywany tylko gdy NIE było wyjątku
finally:
    # wykonywany ZAWSZE (niezależnie od wyniku)
```

| Blok | Kiedy się wykonuje |
|------|-------------------|
| `except` | wyjątek wystąpił |
| `else` | wyjątek NIE wystąpił |
| `finally` | zawsze - do sprzątania zasobów |

> 💡 **Tip:** `else` i `finally` są opcjonalne. W praktyce
> `finally` używamy głównie do zamykania plików i połączeń,
> a `else` - gdy chcemy wyraźnie oddzielić kod "sukcesu"
> od kodu obsługi błędu.

````python
def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError as e:
        print(f"Error: {e}")
        return None
    else:
        print("Division successful")
        return result
    finally:
        print("--- done ---")


print(divide(10, 2))   # else + finally
print(divide(10, 0))   # except + finally
````

## 2. 🔹 Wbudowane wyjątki

Python posiada dziesiątki wbudowanych typów wyjątków. Warto znać
najczęściej spotykane:

| Wyjątek | Kiedy się pojawia |
|---------|------------------|
| `ZeroDivisionError` | dzielenie przez zero |
| `TypeError` | nieodpowiedni typ argumentu |
| `ValueError` | poprawny typ, ale zła wartość |
| `NameError` | odwołanie do niezdefiniowanej zmiennej |
| `IndexError` | indeks poza zakresem listy |
| `KeyError` | brak klucza w słowniku |
| `FileNotFoundError` | plik lub katalog nie istnieje |
| `AttributeError` | obiekt nie ma podanego atrybutu |

Jeden blok `try` może mieć wiele klauzul `except`. Python sprawdza
je po kolei i wykonuje pierwszą pasującą.

> 💡 **Tip:** `except Exception` łapie prawie wszystko, ale ukrywa
> błędy programistyczne. Zawsze wymieniaj konkretne typy wyjątków,
> które faktycznie się spodziewasz.

````python
# Kilka typow wyjatkow naraz
def process(value, data):
    try:
        number = int(value)
        result = data[number]
        return 100 / result
    except ValueError:
        print("Cannot convert to int")
    except IndexError:
        print("Index out of range")
    except ZeroDivisionError:
        print("Division by zero")
    return None


print(process("2",  [10, 20, 0]))   # ZeroDivisionError
print(process("5",  [10, 20, 5]))   # IndexError
print(process("ab", [10, 20, 5]))   # ValueError
````

````python
# Tupla wyjatkow w jednym except
items = {"apple": 3, "banana": 5}

for key in ["apple", "cherry", "banana"]:
    try:
        print(f"{key}: {items[key]}")
    except (KeyError, TypeError) as e:
        print(f"Skipping {key}: {type(e).__name__}")
````

## 3. 🔹 Podnoszenie wyjątków (`raise`)

Słowo kluczowe `raise` służy do *ręcznego* zgłoszenia wyjątku.
Używamy go gdy chcemy wymusić warunki poprawności danych wejściowych
lub sygnalizować błąd logiczny w programie.

```python
raise ValueError("message")   # zgłoś z komunikatem
raise                          # ponownie zgłoś aktywny wyjątek
```

Dobre praktyki:
- wybierz *konkretny* typ wyjątku (nie `Exception` jeśli nie musisz)
- napisz *jasny komunikat* z wartością, która zawodzi
- `raise` bez argumentu w bloku `except` przekazuje wyjątek dalej

````python
# Walidacja argumentow przez raise
def set_age(age):
    if not isinstance(age, int):
        raise TypeError(f"Age must be int, got {type(age).__name__}")
    if age < 0 or age > 150:
        raise ValueError(f"Age {age} is out of range [0, 150]")
    return age


print(set_age(25))    # 25
print(set_age(-1))    # ValueError
````

````python
# register_user - raise + try/except razem
users = ['alice', 'bob', 'carol']


def register_user(username, password):
    if username in users:
        raise ValueError(f"Username '{username}' already taken")
    if len(password) < 8:
        raise ValueError("Password must be at least 8 characters")
    users.append(username)
    print(f"User '{username}' registered successfully")


for username, password in [
    ('dave', 'secret99'),
    ('bob',  'pass1234'),
    ('eve',  'short'),
]:
    try:
        register_user(username, password)
    except ValueError as e:
        print(f"Registration failed: {e}")
````

## 4. 🔹 Własne wyjątki (custom exceptions)

Własny wyjątek tworzymy przez dziedziczenie po klasie `Exception`
(lub jej podklasie). Dzięki temu możemy:
- nadać wyjątkowi *znaczącą nazwę* opisującą problem domenowy
- dodać *dodatkowe atrybuty* (np. kod błędu, wartość, która zawodzi)
- przechwytywać go *selektywnie* niezależnie od wbudowanych wyjątków

```python
class MyError(Exception):
    pass

class MyErrorWithCode(Exception):
    def __init__(self, message, code):
        super().__init__(message)
        self.code = code
```

> 💡 **Tip:** Nazwę własnego wyjątku zawsze kończymy na `Error`
> (np. `InsufficientFundsError`, `ValidationError`).
> Ułatwia to rozpoznanie, że mamy do czynienia z wyjątkiem.

````python
# Prosty wlasny wyjatek
class ValidationError(Exception):
    pass


def validate_username(name):
    if len(name) < 3:
        raise ValidationError(f"Username '{name}' is too short (min 3 chars)")
    if not name.isalnum():
        raise ValidationError(f"Username '{name}' contains invalid characters")
    return name


for name in ['al', 'alice!', 'alice123']:
    try:
        print(validate_username(name))
    except ValidationError as e:
        print(f"Validation failed: {e}")
````

````python
# Wyjatek z atrybutem dodatkowym
class AppError(Exception):
    def __init__(self, message, code=None):
        super().__init__(message)
        self.code = code

    def __str__(self):
        if self.code:
            return f"[{self.code}] {super().__str__()}"
        return super().__str__()


try:
    raise AppError("Resource not found", code=404)
except AppError as e:
    print(e)         # [404] Resource not found
    print(e.code)    # 404
````
