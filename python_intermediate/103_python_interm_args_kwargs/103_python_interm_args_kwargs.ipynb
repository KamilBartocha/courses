## 1. рҹ”№ `*args`

`*args` to konwencja przyjmowania **zmiennej liczby argumentГіw
pozycyjnych**. WewnД…trz funkcji `args` jest dostДҷpne jako **krotka
(tuple)**.

```python
def funkcja(*args):
    # args to krotka wszystkich przekazanych wartoЕӣci
    for arg in args:
        print(arg)
```

| Aspekt | SzczegГіЕӮ |
|---|---|
| Typ wewnД…trz funkcji | `tuple` |
| KolejnoЕӣДҮ w sygnaturze | po zwykЕӮych parametrach |
| Nazwa | konwencja `args`, moЕјe byДҮ dowolna |
| DostДҷp | iteracja, indeksowanie, `len()` |

> рҹ’Ў Gwiazda `*` to operator - `args` to tylko konwencja nazewnicza.
> MoЕјemy napisaДҮ `*numbers`, `*values` - waЕјna jest tylko gwiazdka.

````python
def print_args(*args):
    print(args)
    print(type(args))   # <class 'tuple'>
    for arg in args:
        print(arg)

print_args(1, 2, 3, "hello", [1, 2, 3])
````

````python
def add_all(*args):
    total = 0
    for value in args:
        total += value
    return total

print(add_all(1, 1, 1, 1))       # 4
print(add_all(1, 1, 2, 3, 5, 6)) # 18

# KrГіcej - z sum()
def add_all_v2(*args):
    return sum(args)

print(add_all_v2(10, 20, 30))     # 60
````

````python
def concat_strings(*args):
    return " ".join(args)

print(concat_strings("Hello", "world!"))      # Hello world!
print(concat_strings("Python", "is", "fun"))  # Python is fun
````

## 2. рҹ”№ `**kwargs`

`**kwargs` to konwencja przyjmowania **zmiennej liczby argumentГіw
sЕӮownikowych** (pary klucz-wartoЕӣДҮ). WewnД…trz funkcji `kwargs`
jest dostДҷpne jako **sЕӮownik (dict)**.

```python
def funkcja(**kwargs):
    # kwargs to dict wszystkich przekazanych par klucz=wartoЕӣДҮ
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

| Aspekt | SzczegГіЕӮ |
|---|---|
| Typ wewnД…trz funkcji | `dict` |
| KolejnoЕӣДҮ w sygnaturze | po `*args` |
| Nazwa | konwencja `kwargs`, moЕјe byДҮ dowolna |
| DostДҷp | `.items()`, `.keys()`, `.values()`, `kwargs[key]` |

> рҹ’Ў Do iteracji po `kwargs` uЕјywamy `.items()`, ktГіre zwraca
> pary `(klucz, wartoЕӣДҮ)` - tak samo jak w zwykЕӮym sЕӮowniku.

````python
def print_kwargs(**kwargs):
    print(kwargs)
    print(type(kwargs))   # <class 'dict'>
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_kwargs(name='Alice', age=30, city='New York')
````

````python
def configure_app(**settings):
    defaults = {
        'debug': True,
        'log_level': 'INFO',
        'host': 'localhost',
        'port': 8080,
    }
    defaults.update(settings)   # nadpisujemy wartoЕӣci domyЕӣlne
    return defaults

print(configure_app(debug=False, port=5000))
print(configure_app())
````

## 3. рҹ”№ ЕҒД…czenie `*args` i `**kwargs`

MoЕјemy ЕӮД…czyДҮ zwykЕӮe parametry, `*args` i `**kwargs` w jednej
funkcji. WaЕјna jest **kolejnoЕӣДҮ** w sygnaturze:

```python
def f(param1, param2, *args, kwonly=None, **kwargs):
    pass
```

| KolejnoЕӣДҮ | Typ parametru | PrzykЕӮad |
|---|---|---|
| 1 | wymagane pozycyjne | `a, b` |
| 2 | `*args` | `*args` |
| 3 | keyword-only (po `*args`) | `debug=False` |
| 4 | `**kwargs` | `**kwargs` |

> рҹ’Ў Parametry po `*args` stajД… siДҷ **keyword-only** - moЕјna je
> przekazaДҮ wyЕӮД…cznie przez nazwДҷ: `f(1, 2, debug=True)`.

````python
def example(arg1, arg2, *args, kwarg1=None, **kwargs):
    print(f"arg1={arg1}, arg2={arg2}")
    print(f"args={args}")
    print(f"kwarg1={kwarg1}")
    print(f"kwargs={kwargs}")

example(1, 2, 3, 4, 5,
        kwarg1='value1',
        extra1='extra1', extra2='extra2')
````

````python
def report(title, *args, **kwargs):
    print(f"Title: {title}")
    print("Arguments:", args)
    print("Keyword Arguments:", kwargs)

report('Report 1', 10, 20, 30,
       status='completed', author='John')
````

## 4. рҹ”№ Rozpakowywanie `*` i `**`

Operator `*` i `**` dziaЕӮajД… rГіwnieЕј **poza definicjД… funkcji** -
do rozpakowywania (unpacking) kolekcji.

| UЕјycie | Opis | PrzykЕӮad |
|---|---|---|
| `*list` w wywoЕӮaniu | rozpakuj listДҷ jako args | `f(*[1, 2, 3])` |
| `**dict` w wywoЕӮaniu | rozpakuj dict jako kwargs | `f(**{'a': 1})` |
| `[*a, *b]` | poЕӮД…cz listy | `[*[1,2], *[3,4]]` |
| `{**a, **b}` | poЕӮД…cz/nadpisz dykty | `{**d1, **d2}` |
| `a, *b, c = [...]` | przypisanie z resztД… | `first, *rest, last` |

> рҹ’Ў Przy `{**d1, **d2}` klucze z `d2` **nadpisujД…** klucze z `d1`.
> KolejnoЕӣДҮ ma znaczenie!

````python
# ЕҒД…czenie sЕӮownikГіw przez **
defaults = {'debug': True, 'host': 'localhost', 'port': 8080}
overrides = {'debug': False, 'port': 5000}

config = {**defaults, **overrides}   # overrides wygrywa
print(config)
# {'debug': False, 'host': 'localhost', 'port': 5000}

config2 = {**overrides, **defaults}  # defaults wygrywa
print(config2)
# {'debug': True, 'port': 8080, 'host': 'localhost'}
````

````python
# ЕҒД…czenie list przez *
fruits = ["apple", "banana"]
veggies = ["carrot", "broccoli"]

all_foods = [*fruits, *veggies]
print(all_foods)

# Wstawianie w Еӣrodek
extras = ["cherry", *fruits, "date"]
print(extras)
````

````python
# Przypisanie z * - extended unpacking
numbers = [1, 2, 3, 4, 5]

first, *rest = numbers
print(first)   # 1
print(rest)    # [2, 3, 4, 5]

first, *middle, last = numbers
print(first)    # 1
print(middle)   # [2, 3, 4]
print(last)     # 5
````

````python
# Rozpakowywanie przy wywoЕӮaniu funkcji
def add(a, b, c):
    return a + b + c

values = [1, 2, 3]
print(add(*values))     # 6  - odpowiednik add(1, 2, 3)

params = {'a': 10, 'b': 20, 'c': 30}
print(add(**params))    # 60 - odpowiednik add(a=10, b=20, c=30)
````
