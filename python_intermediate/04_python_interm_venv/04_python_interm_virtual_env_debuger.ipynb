## 1. 🔹 Wirtualne środowisko (`venv`)

**Wirtualne środowisko (virtual environment)** to odizolowana
instalacja Pythona z własnym zestawem pakietów. Każdy projekt
dostaje swoje własne środowisko, niezależne od innych projektów
i od globalnej instalacji Pythona.

| | Bez `venv` | Z `venv` |
|---|---|---|
| Pakiety | globalne, wspólne | lokalne, per projekt |
| Konflikty wersji | możliwe | brak |
| Przenoszalność | trudna | łatwa przez `requirements.txt` |
| Izolacja | brak | pełna |

> 💡 `venv` jest wbudowany w Pythona 3.3+ - nie wymagamy
> dodatkowej instalacji. Używamy `python -m venv`.

### Podstawowe polecenia `venv`

**Tworzenie środowiska:**
```bash
python -m venv nazwa_srodowiska
```

**Aktywacja:**
```bash
# macOS / Linux
source nazwa_srodowiska/bin/activate

# Windows
nazwa_srodowiska\Scripts\activate
```

**Instalacja pakietów (gdy środowisko aktywne):**
```bash
pip install nazwa_pakietu
pip install requests==2.28.0   # konkretna wersja
```

**Dezaktywacja:**
```bash
deactivate
```

Po aktywacji w terminalu pojawia się prefix `(nazwa_srodowiska)`
przed ścieżką - oznacza to, że jesteśmy wewnątrz środowiska.

### Struktura katalogu projektu

```
project_root/
├── venv/                   <- środowisko (nie commitujemy!)
│   ├── bin/
│   ├── lib/
│   │   └── pythonX.Y/
│   │       └── site-packages/
│   └── pyvenv.cfg
├── src/
│   └── main.py
├── requirements.txt        <- commitujemy
├── .gitignore              <- dodajemy venv/ do .gitignore
└── README.md
```

> 💡 Katalogu `venv/` nigdy nie wrzucamy do repozytorium git.
> Dodajemy go do `.gitignore`. Inni deweloperzy odtworzą środowisko
> przez `pip install -r requirements.txt`.

## 2. 🔹 `requirements.txt`

Plik `requirements.txt` to lista pakietów wymaganych przez projekt.
Pozwala odtworzyć identyczne środowisko na innej maszynie.

**Tworzenie `requirements.txt` z aktywnego środowiska:**
```bash
pip freeze > requirements.txt
```

**Instalacja wszystkich pakietów z pliku:**
```bash
pip install -r requirements.txt
```

**Format pliku:**
```
requests==2.31.0
selenium==4.18.0
pytest>=7.0.0
```

| Operator | Znaczenie | Przykład |
|---|---|---|
| `==` | dokładna wersja | `requests==2.31.0` |
| `>=` | co najmniej wersja | `pytest>=7.0.0` |
| `~=` | kompatybilna wersja | `flask~=2.3.0` |
| brak | najnowsza | `requests` |

````python
# Ćw. 6: Wersja pakietu przez importlib.metadata
import importlib.metadata

def get_version(package_name):
    try:
        return importlib.metadata.version(package_name)
    except importlib.metadata.PackageNotFoundError:
        return f"{package_name} not installed"


print(get_version("pip"))
print(get_version("requests"))
print(get_version("nonexistent-package"))
````

## 3. 🔹 Debugger

**Debugger** to narzędzie do analizy kodu krok po kroku. Pozwala
zatrzymać program w dowolnym miejscu i zbadać stan zmiennych.

### Kluczowe operacje

| Operacja | Skrót (VS Code) | Opis |
|---|---|---|
| Breakpoint | `F9` | zatrzymaj program w tej linii |
| Continue | `F5` | kontynuuj do następnego breakpointa |
| Step Over | `F10` | wykonaj linię, nie wchodź do funkcji |
| Step Into | `F11` | wejdź do wnętrza funkcji |
| Step Out | `Shift+F11` | wyjdź z bieżącej funkcji |

W Pythonie możemy też wstawić punkt przerwania programowo:

```python
breakpoint()   # Python 3.7+ - otwiera pdb
```

> 💡 Debugger jest szybszy niż `print()` przy szukaniu błędów.
> Pozwala sprawdzić stan **wszystkich** zmiennych naraz, bez
> modyfikowania kodu.

````python
# Przykład - śledzimy pętlę krok po kroku
word = 'debugging'
letter = 'g'
count = 0

for char in word:
    if char == letter:
        count += 1

print(f"Letter '{letter}' appears {count} times in '{word}'.")

# Umieść breakpoint na linii 'for char in word:' i obserwuj
# jak zmienia się 'char' i 'count' w każdej iteracji.
````

````python
# breakpoint() - wbudowany punkt przerwania (Python 3.7+)
# Odkomentuj breakpoint() i uruchom w terminalu (nie w Jupyter)

def calculate_total(prices):
    total = 0
    for price in prices:
        # breakpoint()   # zatrzymaj tutaj - sprawdź price i total
        total += price
    return total

result = calculate_total([10, 20, 30, 40])
print(f"Total: {result}")
````

### 🐍 Ćwiczenia - Debugger

Poniższe funkcje zawierają błędy. Uruchom je z debuggerem
(lub przez analizę), znajdź i napraw każdy błąd.

**Ćw. 7.** Funkcja ma zwracać sumę liczb od 1 do n.
Wynik dla `n=5` powinien być `15`.

**Ćw. 8.** Funkcja ma zwracać liczbę samogłosek w słowie.
Dla `"hello"` wynik powinien być `2`.

**Ćw. 9. *(Trudniejsze)*** Funkcja ma zwracać listę liczb
pierwszych do `n`. Dla `n=10` wynik: `[2, 3, 5, 7]`.

````python
# Ćw. 7: Znajdź i napraw błąd
def sum_to_n(n):
    total = 0
    for i in range(n):   # błąd tutaj
        total += i
    return total


print(sum_to_n(5))   # powinno być 15, jest: ?
````

````python
# Ćw. 8: Znajdź i napraw błąd
def count_vowels(word):
    vowels = 'aeiou'
    count = 0
    for char in word:
        if char in vowels:   # sprawdź: case-sensitive?
            count == 1       # błąd tutaj
    return count


print(count_vowels("hello"))   # powinno być 2, jest: ?
print(count_vowels("APPLE"))   # powinno być 2, jest: ?
````

````python
# Ćw. 9: Znajdź i napraw błąd (liczby pierwsze)
def primes_up_to(n):
    primes = []
    for num in range(2, n):   # sprawdź: czy n jest wliczone?
        is_prime = True
        for divisor in range(2, num):
            if num % divisor == 0:
                is_prime = False
        if is_prime:
            primes.append(num)
    return primes


print(primes_up_to(10))   # powinno być [2, 3, 5, 7]
print(primes_up_to(11))   # powinno być [2, 3, 5, 7, 11]
````
