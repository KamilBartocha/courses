## 1. 🔹 Faker - podstawy i providery

Biblioteka `faker` generuje realistyczne dane testowe: imiona,
adresy, e-maile, numery kart, daty i wiele innych. Instalujemy:

```bash
pip install faker
```

Tworzymy instancję i wywołujemy metody providerów:

```python

fake = Faker()

fake.email()      # 'alice@example.com'

```

Wybrane providery i ich metody:

| Provider | Przykładowe metody |
|---|---|
| `person` | `name()`, `first_name()`, `last_name()` |
| `internet` | `email()`, `url()`, `ipv4()`, `user_name()` |
| `address` | `address()`, `city()`, `country()`, `postcode()` |
| `date_time` | `date()`, `date_between()`, `date_time()` |
| `company` | `company()`, `job()`, `bs()` |
| `lorem` | `word()`, `sentence()`, `paragraph()` |
| `misc` | `uuid4()`, `boolean()`, `random_int()` |

Locale zmienia język/kraj generowanych danych:


## 2. 🔹 Seed i reprodukowalność testów

Domyślnie `faker` generuje inne dane przy każdym uruchomieniu.
W testach często potrzebujemy deterministycznych danych - takich
samych przy każdym uruchomieniu:

```python
Faker.seed(42)    # seed globalny - wszystkie instancje
fake.seed_instance(42)  # seed lokalny - jedna instancja
```

Po ustawieniu seeda ta sama sekwencja wywołań zwraca te same dane:

```python
Faker.seed(0)
fake = Faker()
name1 = fake.name()   # zawsze to samo
name2 = fake.name()   # zawsze to samo (ale inne niż name1)
```

Seed sprawdza się w testach regresyjnych i snapshotowych,
gdzie porównujemy wynik z oczekiwaną wartością wygenerowaną
raz i zapisaną. Bez seeda test byłby niedeterministyczny.

Metoda `profile()` zwraca słownik z wieloma polami naraz -
przydatna do generowania kompletnych rekordów użytkownika:

```python
fake.profile()  # name, address, birthdate, sex, mail, ...
fake.simple_profile()  # name, mail, birthdate, sex, username
```

````python
from faker import Faker

# Seed globalny - reprodukowalność
Faker.seed(42)
fake = Faker()

run1 = [fake.name() for _ in range(3)]

Faker.seed(42)
fake2 = Faker()
run2 = [fake2.name() for _ in range(3)]

print('run1:', run1)
print('run2:', run2)
print('identyczne:', run1 == run2)

# profile() - kompletny rekord
Faker.seed(0)
fake3 = Faker()
for i in range(3):
    p = fake3.simple_profile()
    print(f"  {p['name']:25} {p['mail']}")
````

````python
import sys, subprocess, tempfile, os

def _run(code, flags=None):
    flags = flags or ['-v', '--tb=short', '-p', 'no:cacheprovider']
    with tempfile.NamedTemporaryFile(
        mode='w', suffix='_test.py', delete=False, dir='.'
    ) as tmp:
        tmp.write(code)
        fname = tmp.name
    result = subprocess.run(
        [sys.executable, '-m', 'pytest', fname] + flags,
        capture_output=True, text=True
    )
    short = os.path.basename(fname)
    print(result.stdout.replace(fname, short)[-2000:])
    os.unlink(fname)

# Ćwiczenie 8: faker + @pytest.mark.parametrize
# Wygeneruj 5 profili przez Faker i użyj ich jako parametrów testu.
# Każdy profil weryfikuje że email zawiera '@' i imię nie jest puste.

_run('''
import pytest
from faker import Faker

Faker.seed(99)
fake = Faker()
profiles = [fake.simple_profile() for _ in range(5)]

@pytest.mark.parametrize("profile", profiles)
def test_profile_has_valid_email(profile):
    assert "@" in profile["mail"]

@pytest.mark.parametrize("profile", profiles)
def test_profile_has_name(profile):
    assert len(profile["name"]) > 0
''')
````

## 3. 🔹 Własny provider i integracja z pytest

Własny provider tworzymy przez dziedziczenie po `BaseProvider`
i rejestrację przez `add_provider()`:

```python
from faker import Faker
from faker.providers import BaseProvider

class ColorProvider(BaseProvider):
    COLORS = ['red', 'green', 'blue', 'yellow']

    def color_name(self):
        return self.random_element(self.COLORS)

fake = Faker()
fake.add_provider(ColorProvider)
print(fake.color_name())  # 'red'
```

Integracja z pytest - fixture `fake` dostępna w całym projekcie:

```python
# conftest.py
import pytest
from faker import Faker

@pytest.fixture
def fake():
    Faker.seed(0)
    return Faker()

@pytest.fixture
def fake_user(fake):
    return {
        'name': fake.name(),
        'email': fake.email(),
        'age': fake.random_int(min=18, max=80),
    }
```

````python
from faker import Faker
from faker.providers import BaseProvider

# Własny provider - polskie numery PESEL
class PeselProvider(BaseProvider):
    def pesel(self):
        """Generuje losowy PESEL (uproszczony, bez walidacji cyfry kontrolnej)."""
        year = self.random_int(min=50, max=99)
        month = self.random_int(min=1, max=12)
        day = self.random_int(min=1, max=28)
        suffix = self.random_int(min=1000, max=9999)
        return f"{year:02d}{month:02d}{day:02d}{suffix}"

    def id_card(self):
        """Numer dowodu osobistego w formacie ABC123456."""
        letters = self.random_letters(length=3)
        digits = self.random_int(min=100000, max=999999)
        return ''.join(letters).upper() + str(digits)


fake = Faker('pl_PL')
fake.add_provider(PeselProvider)

print('Przykładowe dane z własnym providerem:')
for _ in range(4):
    print({
        'name': fake.name(),
        'pesel': fake.pesel(),
        'id_card': fake.id_card(),
    })
````

````python
# Ćwiczenie 9: fixture sesyjna - bank użytkowników *(Trudniejsze)*
# Fixture session-scoped generuje raz 500 użytkowników i udostępnia
# je wszystkim testom w sesji. Napisz testy korzystające z tego banku.

_run('''
import pytest
from faker import Faker

@pytest.fixture(scope="session")
def user_bank():
    Faker.seed(42)
    fake = Faker()
    return [
        {
            "name": fake.name(),
            "email": fake.email(),
            "age": fake.random_int(min=18, max=80),
            "country": fake.country_code(),
        }
        for _ in range(500)
    ]

def test_bank_size(user_bank):
    assert len(user_bank) == 500
    print(f"bank: {len(user_bank)} users")

def test_all_emails_valid(user_bank):
    invalid = [u for u in user_bank if "@" not in u["email"]]
    assert invalid == [], f"{len(invalid)} invalid emails"
    print("all emails valid: OK")

def test_all_adults(user_bank):
    minors = [u for u in user_bank if u["age"] < 18]
    assert minors == []
    ages = [u["age"] for u in user_bank]
    print(f"age range: {min(ages)}-{max(ages)}")

def test_country_codes_valid(user_bank):
    codes = {u["country"] for u in user_bank}
    assert all(len(c) == 2 for c in codes)
    print(f"unique countries: {len(codes)}")
''')
````
