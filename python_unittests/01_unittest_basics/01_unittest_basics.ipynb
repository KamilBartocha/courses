## 1. 🔹 `TestCase` i metody `assert*`

Moduł `unittest` jest wbudowany w Pythona - nie wymaga instalacji.
Testy piszemy jako klasy dziedziczące po `unittest.TestCase`.
Każda metoda testowa musi zaczynać się od `test_`.

```python
import unittest

class TestMyModule(unittest.TestCase):
    def test_something(self):
        result = my_function()
        self.assertEqual(result, expected)
```

Klasa `TestCase` dostarcza bogatego zestawu metod `assert*`:

| Metoda | Sprawdza |
|--------|-----------|
| `assertEqual(a, b)` | `a == b` |
| `assertNotEqual(a, b)` | `a != b` |
| `assertTrue(x)` | `bool(x) is True` |
| `assertFalse(x)` | `bool(x) is False` |
| `assertIsNone(x)` | `x is None` |
| `assertIsNotNone(x)` | `x is not None` |
| `assertIn(a, b)` | `a in b` |
| `assertNotIn(a, b)` | `a not in b` |
| `assertIsInstance(a, b)` | `isinstance(a, b)` |
| `assertAlmostEqual(a, b)` | `round(a-b, 7) == 0` |
| `assertRaises(exc, ...)` | wyjątek `exc` jest zgłoszony |

Kazda metoda `assert*` przyjmuje opcjonalny parametr `msg=`,
ktory wyswietla sie jako dodatkowy komunikat przy niepowodzeniu:

```python
self.assertEqual(result, 7, msg="add(3, 4) should return 7")
```

> 💡 **Tip:** Uzywaj metod `assert*` zamiast gorego `assert` -
> dostarczaja czytelnych komunikatow bledu, np.:
> `AssertionError: 5 != 7` zamiast `AssertionError`.

````python
import unittest


class Calculator:
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

    def multiply(self, a, b):
        return a * b

    def is_positive(self, n):
        return n > 0


class TestCalculator(unittest.TestCase):

    def test_add_two_positive_numbers_returns_sum(self):
        calc = Calculator()
        result = calc.add(3, 4)
        self.assertEqual(result, 7)

    def test_add_negative_numbers_returns_correct_sum(self):
        calc = Calculator()
        self.assertEqual(calc.add(-1, -2), -3)

    def test_is_positive_returns_false_for_zero(self):
        calc = Calculator()
        self.assertFalse(calc.is_positive(0))

    def test_is_positive_returns_true_for_positive(self):
        calc = Calculator()
        self.assertTrue(calc.is_positive(5))

    def test_subtract_non_zero_result_is_not_equal(self):
        calc = Calculator()
        self.assertNotEqual(calc.subtract(10, 3), 0)


# W notebooku uzywamy TextTestRunner zamiast unittest.main()
runner = unittest.TextTestRunner(verbosity=2)
runner.run(unittest.TestLoader().loadTestsFromTestCase(TestCalculator))
````

## 2. 🔹 Uruchamianie i odkrywanie testów

Testy `unittest` mozna uruchamiac na kilka sposobow.

**W notebooku** - uzywamy `TextTestRunner` lub `unittest.main()`
z parametrami blokujacymi wyjscie z kernela:

```python
# Sposob 1 - TextTestRunner (zalecany w notebooku)
runner = unittest.TextTestRunner(verbosity=2)
runner.run(unittest.TestLoader().loadTestsFromTestCase(TestMyClass))

# Sposob 2 - unittest.main z exit=False
if __name__ == '__main__':
    unittest.main(argv=[''], verbosity=2, exit=False)
```

**W terminalu** - uruchamiamy plik bezposrednio lub uzywamy
test discovery (`python -m unittest discover`):

```bash
# Uruchom konkretny plik
python -m unittest test_calculator.py

# Uruchom wszystkie testy w katalogu (test discovery)
python -m unittest discover -v

# Uruchom konkretna klase lub metode
python -m unittest test_calculator.TestCalculator
python -m unittest test_calculator.TestCalculator.test_add
```

**Test discovery** automatycznie wykrywa testy jesli:
- plik ma nazwe pasujaca do wzorca `test*.py` (domyslnie),
- klasa testowa dziedziczy po `unittest.TestCase`,
- metoda testowa zaczyna sie od `test_`.

**`TestSuite`** pozwala laczyc wiele klas testowych
i uruchamiac je razem:

```python
suite = unittest.TestSuite()
suite.addTests(unittest.TestLoader().loadTestsFromTestCase(TestA))
suite.addTests(unittest.TestLoader().loadTestsFromTestCase(TestB))
runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite)
```

> 💡 **Tip:** Parametr `verbosity=2` wyswietla nazwe kazdej
> metody testowej. `verbosity=1` (domyslnie) pokazuje tylko
> `.` (sukces) lub `F` (porazka) dla kazdego testu.

````python
import unittest


# Przyklad: TestSuite laczacy dwie klasy testowe

class StringUtils:
    def reverse(self, text):
        return text[::-1]

    def capitalize_words(self, text):
        return text.title()


class MathUtils:
    def clamp(self, value, lo, hi):
        return max(lo, min(value, hi))

    def sign(self, n):
        if n > 0:
            return 1
        if n < 0:
            return -1
        return 0


class TestStringUtils(unittest.TestCase):
    def test_reverse_hello_returns_olleh(self):
        utils = StringUtils()
        self.assertEqual(utils.reverse("hello"), "olleh")

    def test_capitalize_words_uppercases_each_word(self):
        utils = StringUtils()
        self.assertEqual(utils.capitalize_words("hello world"), "Hello World")


class TestMathUtils(unittest.TestCase):
    def test_clamp_value_below_range_returns_lo(self):
        utils = MathUtils()
        self.assertEqual(utils.clamp(-5, 0, 10), 0)

    def test_sign_returns_minus_one_for_negative(self):
        utils = MathUtils()
        self.assertEqual(utils.sign(-3), -1)


# Lacze dwie klasy w jeden TestSuite
suite = unittest.TestSuite()
suite.addTests(unittest.TestLoader().loadTestsFromTestCase(TestStringUtils))
suite.addTests(unittest.TestLoader().loadTestsFromTestCase(TestMathUtils))

runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite)
````

## 3. 🔹 Testowanie wyjątków i komunikatów

Do testowania wyjatkow sluzy `assertRaises`. Mozemy uzywac go
na dwa sposoby:

**Context manager (zalecany)** - blok `with` przechwytuje
wyjatek zgłoszony wewnatrz niego:

```python
with self.assertRaises(ValueError):
    my_function(-1)
```

**Wywolanie (callable style)** - przekazujemy funkcje
i jej argumenty jako osobne parametry:

```python
self.assertRaises(ValueError, my_function, -1)
```

Aby sprawdzic tresci komunikatu wyjatku, uzywamy
kontekstu jako zmiennej (`as ctx`):

```python
with self.assertRaises(ValueError) as ctx:
    my_function(-1)
self.assertIn("negative", str(ctx.exception))
```

Parametr `msg=` dodaje wlasny komunikat przy niepowodzeniu
assercji - pomaga identyfikowac blad w bardziej zlozonych testach:

```python
self.assertEqual(result, 42, msg="Expected 42 after processing")
```

> 💡 **Tip:** Uzywaj formy context manager - jest czytelniejsza
> i pozwala latwo sprawdzac komunikat wyjatku przez `ctx.exception`.
> Forma callable jest uzyteczna w parametryzowanych testach.

````python
import unittest


class BankAccount:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError(f"Deposit amount must be positive, got {amount}")
        self.balance += amount

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError(f"Withdrawal amount must be positive, got {amount}")
        if amount > self.balance:
            raise ValueError(
                f"Insufficient funds: balance {self.balance}, "
                f"requested {amount}"
            )
        self.balance -= amount


class TestBankAccountExceptions(unittest.TestCase):

    # Context manager - zalecany sposob
    def test_deposit_negative_amount_raises_value_error(self):
        account = BankAccount(balance=100)
        with self.assertRaises(ValueError):
            account.deposit(-50)

    # Sprawdzenie komunikatu wyjatku
    def test_deposit_negative_includes_amount_in_message(self):
        account = BankAccount(balance=100)
        with self.assertRaises(ValueError) as ctx:
            account.deposit(0)
        self.assertIn("positive", str(ctx.exception))

    # Callable style
    def test_withdraw_exceeds_balance_raises_callable_style(self):
        account = BankAccount(balance=50)
        self.assertRaises(ValueError, account.withdraw, 200)


runner = unittest.TextTestRunner(verbosity=2)
runner.run(unittest.TestLoader().loadTestsFromTestCase(
    TestBankAccountExceptions
))
````
