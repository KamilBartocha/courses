## 1. 🔹 Definiowanie funkcji

Funkcja (function) to nazwany blok kodu, który możemy wywołać
wielokrotnie w różnych miejscach programu. Zamiast kopiować ten
sam kod kilkanaście razy, definiujemy go raz i używamy po nazwie.

Do tej pory używaliśmy funkcji wbudowanych jak `print()` czy `len()`.
Teraz nauczymy się tworzyć własne.

```python
def nazwa_funkcji():
    # ciało funkcji - musi być wcięte!
    kod
```

Słowo kluczowe `def` otwiera definicję. Ciało funkcji musi być
wcięte. Funkcję wywołujemy pisząc `nazwa_funkcji()`.

> 💡 Nazywajmy funkcje czasownikiem opisującym co robią:
> `calculate_sum`, `find_max`, `greet_user` - nie `x` ani `func1`.

````python
def greet():
    print("Hello, world!")
    print("Welcome to Python!")


greet()   # pierwsze wywołanie
greet()   # można wywołać wielokrotnie
````

## 2. 🔹 Parametry i argumenty

Parametr (parameter) to nazwa zmiennej w definicji funkcji.
Argument (argument) to wartość przekazywana przy wywołaniu.

```python
def add(x, y):   # x, y to parametry
    ...

add(2, 3)        # 2, 3 to argumenty
```

Możemy też zdefiniować **wartość domyślną** parametru - zostanie
użyta gdy argument nie zostanie przekazany:

```python
def greet(name="World"):
    print(f"Hello, {name}!")
```

> 💡 Parametry z wartością domyślną muszą stać po parametrach
> bez domyślnej. `def f(a, b=1)` jest poprawne,
> `def f(a=1, b)` - nie.

````python
def add(x, y):
    result = x + y
    return result


print(add(2, 3))    # 5
print(add(10, -4))  # 6


# Wartość domyślna
def greet(name="World"):
    print(f"Hello, {name}!")


greet("Alice")  # Hello, Alice!
greet()         # Hello, World!
````

### ⚠️ Pułapka: mutowalny domyślny argument

Domyślne wartości parametrów (default parameter values) są tworzone
**raz** przy definicji funkcji, nie przy każdym wywołaniu. Gdy
domyślną wartością jest obiekt mutowalny (lista, słownik, zbiór),
wszystkie wywołania współdzielą ten sam obiekt.

```python
def f(lst=[]):   # ⚠️ ta lista istnieje przez cały czas życia programu
    lst.append(1)
    return lst
```

Reguła: **nigdy nie używamy mutowalnych obiektów jako domyślnych
argumentów**. Zamiast tego używamy `None` i tworzymy nowy obiekt
wewnątrz funkcji:

```python
def f(lst=None):   # ✅
    if lst is None:
        lst = []
    lst.append(1)
    return lst
```

````python
# ⚠️ PUŁAPKA: mutowalny domyślny argument (mutable default argument)
def add_item_WRONG(item, lst=[]):   # lista tworzona raz przy definicji!
    lst.append(item)
    return lst

print(add_item_WRONG("a"))   # ['a']
print(add_item_WRONG("b"))   # ['a', 'b']  ← nie ['b']!
print(add_item_WRONG("c"))   # ['a', 'b', 'c'] ← lista "pamięta" poprzednie wywołania


# ✅ POPRAWNIE: używamy None jako wartości domyślnej
def add_item_OK(item, lst=None):
    if lst is None:
        lst = []            # nowa lista przy każdym wywołaniu
    lst.append(item)
    return lst

print(add_item_OK("a"))     # ['a']
print(add_item_OK("b"))     # ['b']  ← poprawnie!
print(add_item_OK("c"))     # ['c']
````

## 3. 🔹 Wartość zwracana

Instrukcja `return` kończy wykonywanie funkcji i zwraca wartość
do miejsca wywołania. Funkcja bez `return` zwraca `None`.

| | Bez `return` | Z `return` |
|---|---|---|
| Wynik | Tylko wypisuje | Wartość do użycia |
| Zapis | Nie da się przypisać | `x = funkcja()` |
| Łańcuchowanie | Niemożliwe | `len(add(2, 3))` |

````python
# Bez return - tylko wypisuje, nie zwraca wartości
def print_sum(a, b):
    print(a + b)


# Z return - zwraca wartość do użycia
def calc_sum(a, b):
    return a + b


result = calc_sum(4, 5)
print(f"Result: {result}")      # 9
print(f"Double: {result * 2}")  # 18

x = print_sum(4, 5)  # wypisze 9, ale...
print(x)             # None - bo print_sum nic nie zwraca
````

````python
def fibonacci_numbers(n):
    """Returns Fibonacci numbers less than n."""
    result = []
    a, b = 1, 1
    while a < n:
        result.append(a)
        a, b = b, a + b
    return result


print(fibonacci_numbers(100))
print(fibonacci_numbers(300))
````

## 4. 🔹 Zakres zmiennych - reguła LEGB

Zakres (scope) to obszar programu, w którym zmienna jest dostępna.
Python szuka zmiennej w kolejności LEGB:

- **L**ocal - wewnątrz bieżącej funkcji
- **E**nclosing - w funkcji zewnętrznej (dla zagnieżdżonych)
- **G**lobal - na poziomie modułu (plik `.py`)
- **B**uilt-in - wbudowane nazwy Pythona (`len`, `print`, `range`)

Zmienna lokalna przysłania zmienną globalną o tej samej nazwie.
Słowo kluczowe `global` pozwala zmodyfikować zmienną globalną
wewnątrz funkcji, ale jest rzadko zalecane w praktyce.

> 💡 Dobrą praktyką jest przekazywanie danych przez parametry,
> nie poleganie na zmiennych globalnych. Funkcja bez efektów
> ubocznych jest łatwiejsza do testowania i rozumienia.

````python
x = "global"  # zmienna globalna


def outer():
    x = "enclosing"  # zakres zewnętrznej funkcji

    def inner():
        x = "local"  # zakres lokalny
        print(x)     # local - L w LEGB

    inner()
    print(x)  # enclosing - E w LEGB


outer()
print(x)   # global - G w LEGB


# Modyfikacja zmiennej globalnej (stosuj rzadko!)
counter = 0


def increment():
    global counter
    counter += 1


increment()
increment()
print(counter)  # 2
````

## 5. 🔹 Docstring

Docstring to ciąg dokumentacyjny - tekst opisujący co robi funkcja,
umieszczony w potrójnych cudzysłowach `"""` zaraz po linii `def`.
Jest dostępny przez `help()` i narzędzia do generowania dokumentacji.

```python
def funkcja(arg):
    """Krótki opis co robi funkcja."""
    ...
```

> 💡 Dobry docstring odpowiada na pytanie "co robi funkcja?",
> nie "jak to robi?". Opisujemy zachowanie, nie implementację.

````python
def add(x, y):
    """Returns the sum of x and y."""
    return x + y


def fibonacci_numbers(n):
    """Returns all Fibonacci numbers less than n."""
    result = []
    a, b = 1, 1
    while a < n:
        result.append(a)
        a, b = b, a + b
    return result


help(add)
help(fibonacci_numbers)
````

## 6. 🔹 Rekurencja

Rekurencja (recursion) to technika, w której funkcja wywołuje samą
siebie. Każde wywołanie rozwiązuje mniejszy podproblem, aż osiągnięty
zostanie **warunek bazowy** (base case), który zatrzymuje rekurencję.

| | Iteracja | Rekurencja |
|---|---|---|
| Mechanizm | pętla | funkcja wywołuje siebie |
| Warunek stopu | `while` / `for` | warunek bazowy |
| Czytelność | często prostsza | elegancka dla problemów drzewiastych |
| Ryzyko | brak | `RecursionError` bez warunku bazowego |

> ⚠️ Każda funkcja rekurencyjna MUSI mieć warunek bazowy.
> Bez niego program zgłosi `RecursionError`.

````python
# Iteracja - suma liczb od 0 do n
def sum_iter(n):
    total = 0
    for i in range(n + 1):
        total += i
    return total


# Rekurencja - to samo, bardziej elegancko(?)
def sum_recur(n):
    if n == 0:                  # warunek bazowy!
        return 0
    return n + sum_recur(n - 1)


print(sum_iter(3))   # 6
print(sum_recur(3))  # 6
````

````python
# Wizualizacja wywołań rekurencyjnych
def sum_traced(n, depth=0):
    indent = "  " * depth
    print(f"{indent}sum_traced({n})")
    if n == 0:
        print(f"{indent}  -> returns 0")
        return 0
    result = n + sum_traced(n - 1, depth + 1)
    print(f"{indent}  -> returns {result}")
    return result


sum_traced(3)
````

## 7. 🔹 *args i **kwargs

Czasem nie wiemy z góry, ile argumentów zostanie przekazanych.
Python pozwala to obsłużyć dwoma specjalnymi składniami:

- `*args` - zbiera dodatkowe argumenty **pozycyjne** w krotkę
- `**kwargs` - zbiera dodatkowe argumenty **słownikowe** w słownik

```python
def f(*args, **kwargs):
    print(args)    # krotka: (1, 2, 3)
    print(kwargs)  # słownik: {'x': 10}

f(1, 2, 3, x=10)
```

Kolejność parametrów w definicji musi być:
`def f(stałe, *args, **kwargs)`

> 💡 Nazwy `args` i `kwargs` to tylko konwencja - równie poprawne
> jest `*liczby` czy `**opcje`. Gwiazdki `*` i `**` to składnia,
> nie nazwa.

````python
# *args - dowolna liczba argumentów pozycyjnych
def total(*args):
    print(f"args: {args}")   # krotka
    return sum(args)


print(total(1, 2, 3))        # 6
print(total(10, 20))         # 30


# **kwargs - dowolna liczba argumentów słownikowych
def show_info(**kwargs):
    print(f"kwargs: {kwargs}")  # słownik
    for key, value in kwargs.items():
        print(f"  {key}: {value}")


show_info(name="Alice", age=30, city="Warsaw")


# Mieszane: stałe + *args + **kwargs
def mixed(required, *args, **kwargs):
    print(f"required: {required}")
    print(f"args:     {args}")
    print(f"kwargs:   {kwargs}")


mixed("hello", 1, 2, 3, color="red", size=42)
````
