## 1. 🖨️ `print()` i formatowanie

Funkcja `print()` wyświetla wartości na ekranie. Możemy przekazać
jej dosłowny tekst (w cudzysłowach), zmienną lub wyrażenie.

Trzy popularne sposoby łączenia tekstu ze zmiennymi:

| Sposób | Przykład |
|---|---|
| konkatenacja | `"Wiek: " + str(age)` |
| `.format()` | `"Wiek: {}".format(age)` |
| f-string | `f"Wiek: {age}"` |

> 💡 Zalecam używanie **f-stringów** - są najbardziej czytelne
> i wydajne. Dostępne od Pythona 3.6.

````python
name = "Alice"
age = 25

# trzy sposoby wypisania tego samego
print("Imię: " + name)               # konkatenacja
print("Imię: {}".format(name))       # .format()
print(f"Imię: {name}, wiek: {age}")  # f-string (zalecany)

# print() przyjmuje wiele argumentów - oddziela spacją
print("Wartości:", 1, 2, 3)
print("Suma 2+2 =", 2 + 2)
````

### Starsze sposoby formatowania

Python ma trzy sposoby wstawiania wartości do napisów. Zalecam
f-stringi, ale warto znać starsze metody - są powszechne w cudzym kodzie:

| Sposób | Przykład | Uwagi |
|---|---|---|
| `%` operator | `"%s ma %d lat" % (name, age)` | Python 2 / legacy |
| `.format()` | `"{} ma {} lat".format(name, age)` | Python 2.6+, nadal popularny |
| f-string | `f"{name} ma {age} lat"` | Python 3.6+, **zalecany** |

> 💡  `%` i `.format()` poznajemy by rozumieć istniejący kod - nowego kodu w tych stylach
> nie piszemy.

````python
name = "Alice"
age = 25

# % operator - najstarszy sposób (spotykamy w starym kodzie)
print("Imię: %s, wiek: %d" % (name, age))

# .format() - Python 2.6+, nadal popularny
print("Imię: {}, wiek: {}".format(name, age))
print("Imię: {n}, wiek: {a}".format(n=name, a=age))

# f-string - Python 3.6+, zalecany
print(f"Imię: {name}, wiek: {age}")
````

## 3. 📦 Zmienne i typy danych

Zmienną (variable) tworzymy przez przypisanie wartości operatorem
`=`. Python automatycznie rozpoznaje typ - nie deklarujemy go.

Typy liczbowe:

| Typ | Przykład | Opis |
|---|---|---|
| `int` | `5`, `-3` | liczba całkowita |
| `float` | `3.14` | liczba zmiennoprzecinkowa |
| `complex` | `1 + 2j` | liczba zespolona |

Operatory przypisania skróconego (augmented assignment):

| Skrót | Pełny zapis |
|---|---|
| `x += 5` | `x = x + 5` |
| `x -= 5` | `x = x - 5` |
| `x *= 2` | `x = x * 2` |

> ⚠️ Użycie zmiennej przed przypisaniem powoduje `NameError`.
> Python nie ma niezainicjalizowanych zmiennych jak w C/Java.

````python
# przypisanie - Python sam rozpoznaje typ
counter = 0
price = 19.99
pi = 3.14159

print(type(counter))    # <class 'int'>
print(type(price))      # <class 'float'>

# przypisanie skrócone
counter += 1
print(counter)          # 1

price *= 1.23           # +23% VAT
print(f"{price:.2f}")   # 24.59
````

Python ma dziewięć wbudowanych typów danych. Każda wartość ma swój typ,
który możemy sprawdzić funkcją `type()`.

| Typ | Kategoria | Przykład | Mutowalny? |
|---|---|---|---|
| `int` | Numeryczny | `42`, `-7` | - |
| `float` | Numeryczny | `3.14`, `1.5e-3` | - |
| `complex` | Numeryczny | `1 + 2j` | - |
| `bool` | Logiczny | `True`, `False` | - |
| `str` | Tekstowy | `'hello'` | Nie |
| `list` | Kolekcja | `[1, 2, 3]` | Tak |
| `tuple` | Kolekcja | `(1, 2, 3)` | Nie |
| `set` | Kolekcja | `{1, 2, 3}` | Tak |
| `dict` | Kolekcja | `{'a': 1}` | Tak |

> 💡 **Mutowalny (mutable)** - wartość można zmienić po utworzeniu.
> Listy, zbiory i słowniki są mutowalne. Napisy, krotki i liczby - nie.

````python
# type() zwraca typ danej wartości
print(type(42))        # <class 'int'>
print(type(3.14))      # <class 'float'>
print(type(True))      # <class 'bool'>
print(type('hello'))   # <class 'str'>
print(type([1, 2]))    # <class 'list'>
print(type({'a': 1}))  # <class 'dict'>
````

## 4. ➕ Arytmetyka i operatory

Python obsługuje wszystkie standardowe operatory arytmetyczne.

| Operator | Opis | Przykład | Wynik |
|---|---|---|---|
| `+` | dodawanie | `3 + 2` | `5` |
| `-` | odejmowanie | `3 - 2` | `1` |
| `*` | mnożenie | `3 * 2` | `6` |
| `/` | dzielenie | `5 / 2` | `2.5` |
| `//` | dzielenie całkowite | `5 // 2` | `2` |
| `%` | reszta z dzielenia | `5 % 2` | `1` |
| `**` | potęgowanie | `2 ** 3` | `8` |

> 💡 `/` zawsze zwraca `float`, nawet gdy wynik jest całkowity:
> `4 / 2` → `2.0`. Użyj `//` gdy potrzebujesz liczby całkowitej.

````python
print(3 + 2)    # 5
print(5 / 2)    # 2.5  - zawsze float
print(5 // 2)   # 2    - dzielenie całkowite
print(5 % 2)    # 1    - reszta z dzielenia
print(2 ** 8)   # 256  - potęgowanie

# kolejność działań (matematyczna)
print(3 + 5 * 0.5)    # 5.5  - mnożenie pierwsze
print((3 + 5) * 0.5)  # 4.0  - nawiasy pierwsze
````

### Operatory porównania (comparison operators)

Operatory porównania zwracają `True` lub `False`. Używamy ich głównie
w warunkach `if` i pętlach `while`:

| Operator | Opis | Przykład |
|---|---|---|
| `==` | Równe (equal) | `3 == 3` → `True` |
| `!=` | Różne (not equal) | `3 != 4` → `True` |
| `>` | Większe (greater than) | `5 > 3` → `True` |
| `<` | Mniejsze (less than) | `2 < 4` → `True` |
| `>=` | Większe lub równe | `5 >= 5` → `True` |
| `<=` | Mniejsze lub równe | `3 <= 4` → `True` |

````python
x = 5

print(x == 5)   # True
print(x != 5)   # False
print(x > 3)    # True
print(x < 3)    # False
print(x >= 5)   # True
print(x <= 4)   # False
````

### Operatory logiczne (logical operators)

Operatory logiczne łączą wyrażenia boolowskie:

| Operator | Opis |
|---|---|
| `not x` | Negacja - odwraca wartość logiczną |
| `x and y` | Prawdziwe gdy **oba** wyrażenia są prawdziwe |
| `x or y` | Prawdziwe gdy **przynajmniej jedno** jest prawdziwe |

> 💡 `and`/`or` nie zawsze zwracają `True`/`False` - zwracają jedną z
> wartości: `0 or 'default'` zwróci `'default'` (bo `0` jest "fałszywe").

````python
# not - negacja
print(not True)     # False
print(not False)    # True

# and - prawdziwe gdy oba
age = 20
is_healthy = True
print(age >= 18 and is_healthy)   # True

# or - prawdziwe gdy przynajmniej jedno
is_phone = True
is_pc = False
print(is_phone or is_pc)          # True

# or zwraca pierwszą 'truthy' wartość
result = 0 or 'default'
print(result)   # default

# or zwraca pierwszą 'truthy' wartość
result = 0 or 127
print(result)   # default
````

### Zamiana wartości zmiennych (variable swap)

Python pozwala zamieniać wartości zmiennych w jednej linii - bez zmiennej
pomocniczej:

| Podejście | Kod |
|---|---|
| Inne języki - zmienna pomocnicza | `temp = a; a = b; b = temp` |
| Python - tuple unpacking | `a, b = b, a` |

````python
a = 5
b = 7

# Python: elegancki swap bez zmiennej pomocniczej
a, b = b, a
print(f'a={a}, b={b}')   # a=7, b=5

# Inne języki: wymaga zmiennej pomocniczej (temp)
a = 5
b = 7
temp = a
a = b
b = temp
print(f'a={a}, b={b}')   # a=7, b=5
````

## 5. Typ logiczny, None i konwersja typów

### Typ logiczny (boolean)

`bool` to typ z dokładnie dwiema wartościami: `True` lub `False`. Wyniki
operatorów porównania to zawsze `bool`.

> 💡 Wiele wartości ma naturalne odpowiedniki logiczne. "Falsy" wartości
> (`0`, `0.0`, `''`, `[]`, `{}`, `None`) traktowane są jak `False`,
> a wszystkie pozostałe jak `True`.

````python
print(True, type(True))   # True <class 'bool'>

# Wyniki porównań to bool
print(2 > 8)         # False
print(10 >= 10)      # True

# Falsy vs truthy
print(bool(0))       # False
print(bool(''))      # False
print(bool([]))      # False
print(bool(42))      # True
print(bool('hello')) # True
````

### Typ `None` i `NoneType`

`None` to specjalna wartość Pythona oznaczająca "brak wartości". Jej
typ to `NoneType`. Pojawia się w trzech typowych sytuacjach:

Zawsze porównujemy `None` przez `is`, nie `==`:
```python
if x is None:    # ✅ poprawnie
if x == None:    # ⚠️ działa, ale nie jest zalecane
```

| Sytuacja | Przykład |
|---|---|
| Funkcja bez `return` | `result = print("hello")` → `None` |
| Brakujący klucz w słowniku | `d.get('klucz')` → `None` |
| Jawne "brak wartości" | `x = None` |

> 💡 `None` jest "falsy"
> Uwaga na: `if x is None:` i `if not x:` to nie to samo.

````python
# None to jedyna wartość typu NoneType
x = None
print(x)           # None
print(type(x))     # <class 'NoneType'>
print(bool(None))  # False

````

### `isinstance()` - sprawdzanie typu

`isinstance(var, Typ)` zwraca `True` jeśli `obj` jest oczekiwanego typu:

| Sprawdzenie | `type(x) == int` | `isinstance(x, int)` |
|---|---|---|
| `x = 42` | `True` | `True` |
| `x = True` | `False` | `True` (bool dziedziczy po int) |
| Wiele typów | Niemożliwe prosto | `isinstance(x, (int, float))` |

```python
isinstance(x, None)
```

> 💡 `isinstance()` stosujemy gdy potrzebujemy sprawdzić typ przed
> operacją - np. w walidacji argumentów funkcji.

> uproszczenie `isinstance(object, Class)`!

````python
print(isinstance(42, int))        # True
print(isinstance(3.14, float))    # True
print(isinstance("hi", str))      # True
print(isinstance([1, 2], list))   # True
````

### Konwersja typów (type conversion)

Konwertujemy między typami wbudowanymi funkcjami:

| Funkcja | Konwertuje na | Przykład |
|---|---|---|
| `int(x)` | Liczba całkowita | `int(3.9)` → `3` |
| `float(x)` | Liczba zmiennoprzecinkowa | `float(7)` → `7.0` |
| `str(x)` | Napis | `str(42)` → `'42'` |
| `bool(x)` | Wartość logiczna | `bool(0)` → `False` |

> ⚠️ `int()` **obcina** część ułamkową, nie zaokrągla: `int(3.9)` → `3`.
> Użyj `round()`, żeby zaokrąglić do najbliższej liczby całkowitej.

````python
# int → float
x = 7
print(float(x))   # 7.0

# float → int (obcina, nie zaokrągla!)
y = 3.99
print(int(y))     # 3  ← obcina
print(round(y))   # 4  ← zaokrągla

# str → int
s = '42'
print(int(s), type(int(s)))   # 42 <class 'int'>

# int → str (wymagane do konkatenacji z napisem)
age = 25
print('Age: ' + str(age))   # Age: 25
````

Konwersja nie zawsze jest dozwolona.
> ValueError: invalid literal for int() with base 10: 'Alice'

````python
x = "Alice"
y = int(x)
````

## 6. 📝 Łańcuchy znaków (strings)

Łańcuch znaków (string) to ciąg znaków w cudzysłowach `""` lub
apostrofach `''`. Python nie rozróżnia między nimi.

Znaki specjalne (escape sequences):

| Sekwencja | Znaczenie |
|---|---|
| `\n` | nowa linia |
| `\t` | tabulacja |
| `\\` | dosłowny backslash |
| `\'` | dosłowny apostrof |

> ⚠️ Stringi są **niezmienne (immutable)** - nie modyfikujemy
> pojedynczego znaku wewnątrz istniejącego str.

````python
greeting = "Hello, World!"
name = 'Alice'

# konkatenacja(dodawanie) i mnożenie str
print("Hello, " + name + "!")
print("-" * 30)

# znaki specjalne
print("Linia 1\nLinia 2")   # \n = nowa linia
print("Kol 1\tKol 2")       # \t = tabulacja
````

Stringi możemy **indeksować (indexing)** - pierwszy znak ma indeks
`0`, ostatni `-1`.

```
 +---+---+---+---+---+---+
 | P | y | t | h | o | n |
 +---+---+---+---+---+---+
   0   1   2   3   4   5
  -6  -5  -4  -3  -2  -1
```

**Wycinanie (slicing)**: `s[start:stop]` - stop jest wyłączony.

| Wyrażenie | Wynik | Opis |
|---|---|---|
| `s[0]` | `'P'` | pierwszy znak |
| `s[-1]` | `'n'` | ostatni znak |
| `s[0:2]` | `'Py'` | indeksy 0, 1 |
| `s[2:]` | `'thon'` | od 2 do końca |
| `s[:3]` | `'Pyt'` | od początku do 3 |
| `s[::-1]` | `'nohtyP'` | odwrócony string |

````python
word = "Python"

print(word[0])       # P   - pierwszy znak
print(word[-1])      # n   - ostatni znak
print(word[0:2])     # Py
print(word[2:])      # thon
print(word[:3])      # Pyt
print(word[::-1])    # nohtyP - krok -1 = od końca

print(len(word))     # 6 - długość (length)

# niezmienność - budujemy nowy string
new_word = "J" + word[1:]
print(new_word)      # Jython
````

### Metody łańcuchów znaków (string methods)

| Metoda | Co robi | Przykład |
|---|---|---|
| `len(s)` | Długość napisu | `len('hello')` → `5` |
| `s.upper()` | Wielkie litery | `'abc'.upper()` → `'ABC'` |
| `s.lower()` | Małe litery | `'ABC'.lower()` → `'abc'` |
| `s.strip()` | Usuwa białe znaki z brzegów | `'  hi  '.strip()` → `'hi'` |
| `s.replace(a, b)` | Zastępuje `a` przez `b` | `'aXb'.replace('X', '-')` → `'a-b'` |
| `s.split(sep)` | Dzieli na listę | `'a,b'.split(',')` → `['a', 'b']` |
| `s.count(sub)` | Zlicza wystąpienia | `'hello'.count('l')` → `2` |
| `s.startswith(p)` | Sprawdza prefix | `'Python'.startswith('Py')` → `True` |

````python
text = 'Hello, World!'

print(len(text))                 # 13
print(text.upper())              # HELLO, WORLD!
print(text.lower())              # hello, world!
print(text.replace('o', '0'))    # Hell0, W0rld!
print(text.count('l'))           # 3
print(text.split(', '))          # ['Hello', 'World!']
print(text.startswith('Hello'))  # True

s = '   python   '
print(s.strip())   # python
````

Dodajemy `r` przed stringiem aby wyłączyć interpretację sekwencji:

`r"C:\Users\name"` wypisze dosłownie `C:\Users\name` - omijamy działanie \

````python
# raw string - backslash nie jest interpretowany
path_no_r = "C:\name"
print(path_no_r)


path = r"C:\name"
print(path)
````

## 7. ⌨️ Dane wejściowe - `input()`

Funkcja `input()` zatrzymuje program i czeka, aż użytkownik wpisze
tekst i naciśnie Enter. Zwraca zawsze **string** - nawet gdy
użytkownik wpisze liczbę.

Aby użyć wartości jako liczby, musimy jawnie skonwertować typ:

| Konwersja | Przykład | Wynik |
|---|---|---|
| `int(input())` | `"42"` → `42` | liczba całkowita |
| `float(input())` | `"3.14"` → `3.14` | liczba zmiennoprzecinkowa |
| `str(input())` | `"Alice"` → `"Alice"` | string (domyślny) |

> ⚠️ Jeśli użytkownik wpisze tekst tam, gdzie oczekujemy liczby,
> `int()` lub `float()` zgłosi `ValueError`. Obsługę błędów
> poznamy w module o wyjątkach.

````python
# input() zawsze zwraca string
name = input("Podaj imię: ")
print(name)

# konwersja na int aby wykonać obliczenia
age = int(input("Podaj wiek: "))
print(f"Za 10 lat będziesz mieć {age + 10} lat.")
````
