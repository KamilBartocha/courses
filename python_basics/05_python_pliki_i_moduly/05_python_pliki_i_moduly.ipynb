## 1. 🔹 Otwieranie pliku (opening a file)

Zanim zaczniemy czytać lub zapisywać dane, musimy otworzyć plik. Służy do
tego wbudowana funkcja `open()` - zwraca obiekt pliku (file object), przez
który możemy operować na jego zawartości.

```python
open('nazwa_pliku.ppp', 'tryb')  # → file object
```

Tryb (mode) mówi Pythonowi, co chcemy z plikiem zrobić:

| Tryb | Działanie |
|---|---|
| `'r'` | odczyt (read) - domyślny, błąd gdy plik nie istnieje |
| `'w'` | zapis (write) - tworzy nowy lub nadpisuje istniejący |
| `'a'` | dopisanie (append) - dodaje na koniec pliku |

> 💡 Każdy otwarty plik zajmuje zasoby systemowe. Zawsze go zamykamy -
> `f.close()`. Możemy też użyć `with` (sekcja 4), które zrobi to automatycznie.

````python
# Opening a file in read mode
f = open('05_file_text.txt', 'r')
print(f.name)   # name of the file
print(f.mode)   # opening mode: 'r'
f.close()

# Opening in write mode (creates the file if it doesn't exist)
f = open('output.txt', 'w')
print(f.name)
print(f.mode)   # 'w'
f.close()
````

## 2. 🔹 Czytanie pliku (reading a file)

Po otwarciu pliku mamy kilka sposobów na odczytanie jego zawartości:

| Metoda (method) | Zwraca | Kiedy używać |
|---|---|---|
| `f.read()` | cały plik jako `str` | małe pliki |
| `f.readline()` | jedną linię jako `str` | linia po linii |
| `f.readlines()` | lista (list) wszystkich linii | gdy potrzebujemy indeksów |
| pętla `for line in f` | jedna linia na iterację | duże pliki |

> 💡 Po `f.read()` kursor (cursor) jest na końcu pliku. Kolejne `f.read()`
> zwróci pusty string. Zamykamy i otwieramy ponownie albo używamy `f.seek(0)`,
> by wrócić na początek.

````python
# f.read() - entire file as one string
f = open('05_file_text.txt', 'r')
content = f.read()
print(content)
f.close()

# f.readline() - one line at a time
f = open('05_file_text.txt', 'r')
first_line = f.readline()
print('First line:', first_line)
f.close()

# f.readlines() - all lines as a list
f = open('05_file_text.txt', 'r')
lines = f.readlines()
print(f'Number of lines: {len(lines)}')
print('Last line:', lines[-1])
f.close()

# for loop - most memory-efficient
f = open('05_file_text.txt', 'r')
for line in f:
    print(line.strip())
f.close()
````

## 3. 🔹 Zapis do pliku (writing to a file)

Do zapisywania danych służy metoda (method) `f.write()`. Przyjmuje wyłącznie
typ `str` - liczby musimy najpierw skonwertować przez `str()`.

```python
f.write(text)   # zapisuje string, zwraca liczbę zapisanych znaków
```

Różnica między trybami (modes):
- `'w'` - zapis (write): tworzy nowy plik lub **nadpisuje** istniejący
- `'a'` - dopisanie (append): dodaje na **koniec** istniejącego pliku

⚠️ Tryb `'w'` usuwa całą poprzednią zawartość pliku - używamy ostrożnie!

````python
# Writing - 'w' mode creates or overwrites
f = open('output.txt', 'w')
f.write('Hello, Python!\n')
f.write('Second line\n')
f.close()

# Appending - 'a' mode adds to the end
f = open('output.txt', 'a')
f.write('Third line - appended\n')
f.close()

# Verify: read back the file
f = open('output.txt', 'r')
print(f.read())
f.close()
````

## 4. 🔹 Menadżer kontekstu (context manager)

Instrukcja (statement) `with` automatycznie zamyka plik po wyjściu z bloku -
nawet jeśli wystąpi błąd wewnątrz. To zalecany sposób pracy z plikami w
Pythonie.

```python
with open(nazwa_pliku, tryb) as f:
    ...   # operujemy na pliku tutaj
# plik jest automatycznie zamknięty po wyjściu z bloku
```

| | Bez `with` | Z `with` |
|---|---|---|
| Zamykanie pliku | ręczne `f.close()` | automatyczne |
| Obsługa błędów | wymaga `try/finally` | wbudowana |
| Czytelność | niska | wysoka |

> 💡 Menadżer kontekstu (context manager) to wzorzec zapewniający poprawne
> zarządzanie zasobami. Działa z każdym obiektem posiadającym metody
> `__enter__` i `__exit__` - nie tylko z plikami.

````python
# ── BEZ with (without context manager) ──────────────────────────────────
f = open('05_file_text.txt', 'r')
try:
    content = f.read()
finally:
    f.close()   # must remember to close - even if error occurs

# ── Z with (with context manager) ────────────────────────────────────────
with open('05_file_text.txt', 'r') as f:
    content = f.read()
# file is automatically closed here

print(content[:80])   # print first 80 characters
````

## 5. 🔹 Moduły (modules)

Moduł (module) to plik Pythona zawierający funkcje, klasy lub zmienne, które
możemy importować (import) do naszego programu. Dzięki modułom nie musimy
pisać wszystkiego od zera.

Python dostarczany jest z bogatą **biblioteką standardową (standard library)**
- zestawem gotowych modułów dostępnych w każdej instalacji. Popularny slogan:
**"Batteries Included"**.

Sposoby importowania:

```python
import modul                   # importuj cały moduł
import modul as alias          # importuj pod aliasem (alias)
from modul import funkcja      # importuj konkretną funkcję
```

⚠️ Unikajmy `from modul import *` - może powodować kolizje nazw (name
collisions) i utrudnia czytanie kodu.

````python
import math                        # import entire module
print(math.pi)
print(math.sqrt(16))

# ── with alias ────────────────────────────────────────────────────────────
import math as m                   # shorter alias
print(m.floor(3.7))
print(m.ceil(3.2))

# ── specific function ─────────────────────────────────────────────────────
from math import sqrt, pi          # import only what you need
print(sqrt(25))
print(pi)

# ── standard library examples ─────────────────────────────────────────────
import os
print(os.getcwd())                 # current working directory

from time import sleep
print('waiting 1 second...')
sleep(1)
print('done')
````

## 7. 🔹 Wirtualne środowiska (virtual environments)

Problem: różne projekty mogą wymagać różnych wersji tych samych bibliotek.
Wirtualne środowisko (virtual environment) to izolowana kopia Pythona
z własnym zestawem pakietów - każdy projekt dostaje swoje "pudełko".

```sh
# Tworzenie środowiska (w katalogu projektu)
python -m venv venv

# Aktywacja
# Windows:
venv\Scripts\activate
# macOS / Linux:
source venv/bin/activate

# Instalacja pakietów tylko w tym środowisku
pip install requests pandas

# Zapisanie listy zależności
pip freeze > requirements.txt

# Instalacja z pliku (np. na nowym komputerze)
pip install -r requirements.txt

# Dezaktywacja środowiska
deactivate
```

Plik `requirements.txt` wygląda tak:
```
requests==2.31.0
pandas==2.1.0
```

> 💡 Katalog `venv/` dodajemy do `.gitignore` - nie commitujemy go do
> repozytorium. Commitujemy tylko `requirements.txt`, żeby inni mogli
> odtworzyć środowisko komendą `pip install -r requirements.txt`.

````python
import sys
import os

# Sprawdzamy czy działamy w środowisku wirtualnym
in_venv = sys.prefix != sys.base_prefix
print(f"W środowisku wirtualnym: {in_venv}")
print(f"Interpreter: {sys.executable}")
print(f"Prefix:      {sys.prefix}")

# Wyświetlamy zainstalowane pakiety
import subprocess
result = subprocess.run(['pip', 'list', '--format=columns'],
                       capture_output=True, text=True)
lines = result.stdout.strip().split('\n')
print(f"\nZainstalowane pakiety: {len(lines) - 2}")  # -2 = nagłówki
for line in lines[:6]:
    print(line)
````

## 8. 🔹 `__name__ == "__main__"`

Każdy plik Pythona ma specjalną zmienną `__name__`. Jej wartość zależy od
tego, jak plik został uruchomiony:

| Sposób | Wartość `__name__` |
|---|---|
| Plik uruchomiony bezpośrednio (`python skrypt.py`) | `'__main__'` |
| Plik zaimportowany przez inny moduł (`import skrypt`) | `'skrypt'` |

Dzięki temu możemy umieścić w pliku kod, który uruchomi się tylko przy
bezpośrednim wywołaniu - nie przy imporcie:

```python
def main():
    print("Program działa!")

if __name__ == '__main__':
    main()
```

To standardowy wzorzec w każdym skrypcie Pythona, który może być
zarówno uruchamiany, jak i importowany.

> 💡 Dobrą praktyką jest zamykanie całego kodu "wykonywalnego" w funkcji
> `main()` i uruchamianie jej przez `if __name__ == '__main__':`.
> Ułatwia to testowanie i ponowne użycie modułu.

````python
# Symulacja: co widzi Python gdy uruchamia plik vs importuje go

# Gdy uruchamiamy plik bezpośrednio:
#   __name__ == '__main__'
# Gdy plik jest importowany jako moduł:
#   __name__ == 'nazwa_pliku' (bez .py)

print(f"__name__ tego notebooka: {__name__}")
# W Jupyterze wartość to '__main__' - tak samo jak przy bezpośrednim uruchomieniu


# Typowy wzorzec w skryptach .py:
#
# def main():
#     print("Program działa!")
#
# if __name__ == '__main__':
#     main()
#
# Dzięki temu: uruchamiamy plik → main() się wywołuje
#              importujemy plik  → main() się NIE wywołuje
````

````python
# Ćwiczenie 3 - calculator.py z wzorcem main()
with open('calculator.py', 'w') as f:
    f.write('def add(a, b):\n')
    f.write('    return a + b\n\n')
    f.write('def subtract(a, b):\n')
    f.write('    return a - b\n\n')
    f.write('def main():\n')
    f.write('    print(f"3 + 5 = {add(3, 5)}")\n')
    f.write('    print(f"10 - 4 = {subtract(10, 4)}")\n\n')
    f.write('if __name__ == "__main__":\n')
    f.write('    main()\n')

import importlib
import calculator
importlib.reload(calculator)        # przeładuj jeśli już importowany
print(calculator.add(..., ...))     # hint: użyj add() z dwoma liczbami
print(calculator.subtract(..., ...))# hint: użyj subtract() z dwoma liczbami
````
