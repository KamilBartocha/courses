## 1. 🔹 Czym jest OOP?

Bez programowania obiektowego (OOP) powiązane dane o jednej rzeczy (imię kota,
wiek, rasa) przechowujemy w osobnych zmiennych, a funkcje działające na nich są
zdefiniowane zupełnie gdzie indziej - dane i zachowanie są od siebie oderwane.
OOP rozwiązuje ten problem, łącząc je w jedną jednostkę: **obiekt (object)**.

Szablon opisujący, jak ma wyglądać obiekt, nazywamy **klasą (class)**. Każdy
obiekt stworzony na podstawie tej klasy nazywamy **instancją (instance)**.

| Pojęcie | Analogia | W kodzie |
|---|---|---|
| Klasa (class) | Przepis / foremka na ciastka | `class Dog:` |
| Obiekt (object/instance) | Konkretne ciastko | `Dog("Rex")` |
| Atrybut (attribute) | Właściwość obiektu | `rex.name` |
| Metoda (method) | Coś, co obiekt potrafi zrobić | `rex.bark()` |



````python
# ── WITHOUT OOP ──────────────────────────────────────────────────────────
cat_name  = "Gatsby"
cat_age   = 2
cat_breed = "Devon Rex"

def cat_speak(name):
    print(f"{name}: Meow!")

cat_speak(cat_name)   # data and behavior are separated

# ── WITH OOP ─────────────────────────────────────────────────────────────
class Cat:
    def __init__(self, name, age, breed):
        self.name  = name
        self.age   = age
        self.breed = breed

    def speak(self):
        print(f"{self.name}: Meow!")

my_cat = Cat("Gatsby", 2, "Devon Rex")
my_cat.speak()        # data and behavior live together
````

## 2. 🔹 Wszystko jest obiektem (object)

W Pythonie *każda wartość* - liczby całkowite (integers), łańcuchy znaków
(strings), listy (lists), a nawet funkcje (functions) - jest obiektem (object).
Oznacza to, że każda wartość ma typ (type), wartość (value) i tożsamość
(identity). Zrozumienie tego wyjaśnia wiele zachowań Pythona.

Zmienne (variables) w Pythonie to **etykiety** wskazujące na obiekty, nie
pudełka przechowujące wartości. Gdy piszesz `x = 4`, przyczepiasz etykietę `x`
do obiektu całkowitego (integer object) o wartości `4`. Przypisanie `x =
'hello'` przenosi etykietę na inny obiekt - liczba `4` pozostaje niezmieniona.

> 💡 Właśnie dlatego możesz przypisać zmiennej wartość zupełnie innego typu -
> typ należy do *obiektu*, nie do *nazwy zmiennej*.


````python
x = 4
print(type(x))      # <class 'int'>

x = 'hello'
print(type(x))      # <class 'str'>

x = 3.14159
print(type(x))      # <class 'float'>

x = [1, 2, 3]
print(type(x))      # <class 'list'>

# Even a function is an object!
def greet(): pass
print(type(greet))  # <class 'function'>
````

## 3. 🔹 Tożsamość (identity), typ (type) i wartość (value)

Każdy obiekt (object) w Pythonie ma trzy cechy:

| Cecha | Co to jest | Jak to sprawdzić |
|---|---|---|
| **Tożsamość (identity)** | Unikalny adres w pamięci | `id(obj)` |
| **Typ (type)** | Klasa, do której należy obiekt | `type(obj)` |
| **Wartość (value)** | Dane przechowywane w obiekcie | `print(obj)` |

**Wbudowana funkcja `id()`**
```python
id(object)  # → int
```
Zwraca unikalną liczbę całkowitą identyfikującą obiekt przez cały czas jego
istnienia. W CPython (standardowy Python) jest to adres obiektu w pamięci.

> 💡 Dwie zmienne z takim samym `id()` wskazują na *ten sam* obiekt w pamięci.
> Operator `is` sprawdza tożsamość (identity) (`a is b`), natomiast `==`
> sprawdza równość wartości (value equality).

⚠️ **Częsty błąd początkujących:** małe liczby całkowite (integers) od `-5` do
`256` są przez Pythona buforowane (cached) - zawsze mają to samo `id()`. Nie
używaj `is` do porównywania wartości.


````python
x = 4
print(f"id={id(x)}, type={type(x)}, value={x}")

x = 'hello'
print(f"id={id(x)}, type={type(x)}, value={x}")

# Two variables pointing at the same object
a = [1, 2, 3]
b = a           # b is an alias - same object!
print(id(a) == id(b))   # True
print(a is b)           # True

# Two variables with equal values but different objects
a = [1, 2]
b = a[:]        # b is a copy
print(id(a) == id(b))   # False
print(a == b)           # True  (same value)
print(a is b)           # False (different objects)
````

## 4. 🔹 Atrybuty (attributes) i metody (methods)

Prawie wszystko, co robimy w Pythonie, polega na wywoływaniu metody (method)
lub odczytywaniu atrybutu (attribute) obiektu. Zrozumienie tej różnicy sprawia,
że notacja z kropką (`obiekt.coś`) staje się intuicyjna.

Obiekty mają dwa rodzaje rzeczy do nich przyczepionych:
- **Atrybuty (attributes)** - *dane* przechowywane na obiekcie (jak zmienna,
  która do niego należy)
- **Metody (methods)** - *funkcje* należące do obiektu (mogą odczytywać i
  modyfikować jego dane)

Dostęp do obu uzyskujemy przez **notację z kropką (dot notation)**:
`obiekt.nazwa`.

```python
x = 4.5
x.real          # atrybut (attribute) - część rzeczywista liczby zespolonej
x.is_integer()  # metoda (method) - musi być wywołana z ()
```

> 💡 Różnica jest prosta: jeśli potrzebuje `()` żeby działać - to metoda
> (method). Jeśli tylko odczytujemy wartość - to atrybut (attribute).

````python
# ── Attributes ───────────────────────────────────────────────────────────
x = 4 + 3j          # complex number
print(x.real)       # attribute: real part
print(x.imag)       # attribute: imaginary part

# ── Methods ──────────────────────────────────────────────────────────────
y = 4.0
print(y.is_integer())    # method: checks if the float has no fractional part

text = "hello world"
print(text.upper())      # method: returns uppercase string
print(text.split())      # method: splits on whitespace, returns a list
print(text.replace("world", "Python"))  # method: string substitution

lst = [3, 1, 4, 1, 5]
lst.append(9)            # method: modifies the list in place
print(lst)
````

## 5. 🔹 Funkcje (functions) vs. metody (methods)

**Po co mi to?** Python ma zarówno wbudowane *funkcje (built-in functions)*
(`sorted()`, `len()`) jak i *metody obiektów (object methods)* (`.sort()`,
`.append()`). Wyglądają podobnie, ale zachowują się zupełnie inaczej -
pomylenie ich prowadzi do błędów i nieoczekiwanych wartości `None`.

| | Funkcja (function) | Metoda (method) |
|---|---|---|
| Składnia (syntax) | `sorted(my_list)` | `my_list.sort()` |
| Należy do | Modułu / wbudowanych | Obiektu |
| Modyfikuje oryginał? | Nie, nowy obiekt | Tak, w miejscu (in place) |
| Działa na każdym typie? | Tylko obsługiwane typy | Tylko własny typ |

> 💡 Metoda (method) to po prostu funkcja (function) należąca do obiektu. Gdy
> Python uruchamia `obj.method()`, automatycznie przekazuje `obj` jako pierwszy
> argument.

⚠️ **Klasyczny błąd:** `my_list = my_list.sort()` - `.sort()` zwraca `None`!
Jeśli chcesz otrzymać nową posortowaną listę, użyj `sorted(my_list)`.


````python
numbers = [6, 9, 3, 1]

# sorted() - FUNCTION: returns a new list, original unchanged
new_list = sorted(numbers)
print("sorted():", new_list)
print("original:", numbers)   # unchanged
print("id same?", id(numbers) == id(new_list))  # False - new object

print()

# .sort() - METHOD: modifies in place, returns None
numbers2 = [6, 9, 3, 1]
result = numbers2.sort()
print(".sort() return value:", result)    # None!
print("numbers2 after .sort():", numbers2) # modified
````

## 6. 🔹 Tworzenie własnych klas (classes)

Do tej pory korzystaliśmy z obiektów (objects) dostarczonych przez Pythona.
Teraz sami je stworzymy. To fundament każdego poważniejszego programowania w
Pythonie.

### Składnia klasy (class syntax)
```python
class NazwaKlasy:              # definicja klasy (class definition)
    def __init__(self, ...):   # konstruktor - uruchamia się przy tworzeniu
        self.atrybut = value   # atrybut instancji (instance attribute)

    def metoda(self):          # metoda instancji (instance method)
        ...
```

### `__init__` - konstruktor (constructor)
`__init__` jest wywoływany automatycznie gdy piszemy `NazwaKlasy(...)`. Ustawia
on początkowy stan (initial state) obiektu.

### `self` - sam obiekt
`self` jest automatycznie przekazywany jako pierwszy argument do każdej metody
(method). Wskazuje na *konkretną instancję (instance)*, na której wywoływana
jest metoda. Właśnie dlatego `obj1.method()` i `obj2.method()` działają każdy
na swoich własnych danych.

> 💡 `self.name` to *atrybut instancji (instance attribute)* - każdy obiekt ma
> własną kopię. Zwykła zmienna `name` wewnątrz metody to *zmienna lokalna
> (local variable)*, która znika po zakończeniu metody.

````python
class Cat:
    def __init__(self, name, age):
        self.name = name   # instance attribute
        self.age  = age    # instance attribute

    def show(self):
        print(f"I am {self.name} and I am {self.age} years old.")

    def speak(self):
        print(f"{self.name}: Meow!")


# Creating instances
a1 = Cat("Devon", 2)
a2 = Cat("Ragdoll", 4)
a3 = Cat("Filemon", 10)

a1.show()
a2.show()
a3.speak()
print(a1.name, a2.age)   # read attributes directly
````

````python
# Visualising self - each instance has its own attribute space
class MyClass:
    def __init__(self, value):
        self.value = value   # stored on THIS instance

    def show_value(self):
        print(f"My value is: {self.value}")


obj1 = MyClass(10)
obj2 = MyClass(20)

# self in obj1.show_value() refers to obj1
# self in obj2.show_value() refers to obj2
obj1.show_value()   # My value is: 10
obj2.show_value()   # My value is: 20

# Proof: they are separate objects in memory
print(id(obj1), id(obj2))   # different IDs
````

## 7. 🔹 Cztery filary OOP

OOP opiera się na czterech zasadach. W tym kursie poznaliśmy ich podstawy -
oto krótki przegląd każdej z nich.

- **Enkapsulacja (encapsulation)** - ukryj szczegóły wewnętrzne,
  udostępnij czysty interfejs
- **Dziedziczenie (inheritance)** - buduj nowe klasy na podstawie
  istniejących
- **Polimorfizm (polymorphism)** - różne obiekty reagują na tę samą
  metodę na swój sposób
- **Abstrakcja (abstraction)** - pracuj z obiektami na wysokim poziomie,
  bez znajomości szczegółów

````python
# ── Enkapsulacja (encapsulation) ─────────────────────────────────────────
class Thermometer:
    def __init__(self, celsius):
        self.__celsius = celsius       # hidden detail

    def temperature(self):
        return f"{self.__celsius}°C"  # clean interface

t = Thermometer(22)
print(t.temperature())               # use the interface, not __celsius directly

# ── Dziedziczenie (inheritance) ──────────────────────────────────────────
class Animal:
    def __init__(self, name, a_type="Animal"):
        self.name = name
        self.a_type = a_type

class Dog(Animal):                   # Dog reuses Animal's __init__
    def speak(self):
        print(f"{self.name} ({self.a_type}): Woof!")

class Cat(Animal):                   # Cat reuses Animal's __init__ too
    def speak(self):
        print(f"{self.name} ({self.a_type}): Meow!")

# ── Polimorfizm (polymorphism) ───────────────────────────────────────────
pets = [Dog("Rex"), Cat("Whiskers")]
for pet in pets:
    pet.speak()                      # same call, different output

# ── Abstrakcja (abstraction) ─────────────────────────────────────────────
numbers = [3, 1, 4, 1, 5]
numbers.sort()                       # no need to know HOW sorting works
print(numbers)
````

## 8. 🔹 Dziedziczenie (inheritance)

Dziedziczenie pozwala tworzyć nowe klasy na podstawie istniejących. Klasa
potomna (child class) przejmuje atrybuty i metody klasy bazowej (base class /
parent class) i może je rozszerzać lub nadpisywać.

```python
class Child(Parent):
    def __init__(self, ...):
        super().__init__(...)  # inicjalizacja rodzica
        ...                    # własne atrybuty
```

Kluczowe pojęcia:

| Pojęcie | Opis |
|---|---|
| `class Dog(Animal):` | `Dog` dziedziczy po `Animal` |
| `super()` | referencja do klasy rodzica |
| `super().__init__(...)` | wywołuje konstruktor rodzica |
| Nadpisanie metody (method overriding) | klasa potomna redefiniuje metodę rodzica |
| `isinstance(obj, Klasa)` | sprawdza czy obiekt należy do klasy lub jej podklas |

Dziedziczenie realizuje zasadę **"is-a"**: `Dog` is-a `Animal`.
Klasa potomna dziedziczy wszystkie metody rodzica i może dodawać własne
lub nadpisywać istniejące.

> 💡 `super().__init__()` musi być wywołane gdy klasa potomna ma własny
> `__init__`, żeby atrybuty rodzica zostały poprawnie ustawione.

````python
class Animal:
    def __init__(self, name, sound):
        self.name  = name
        self.sound = sound

    def speak(self):
        print(f"{self.name}: {self.sound}!")

    def describe(self):
        print(f"I am {self.name}")


class Dog(Animal):
    def __init__(self, name):
        super().__init__(name, "Woof")  # wywołujemy __init__ rodzica

    def fetch(self):
        print(f"{self.name} fetches the ball!")


class Cat(Animal):
    def __init__(self, name):
        super().__init__(name, "Meow")

    def speak(self):  # nadpisujemy metodę rodzica (method overriding)
        print(f"{self.name}: Purrrr... Meow!")


rex    = Dog("Rex")
whiskers = Cat("Whiskers")

rex.speak()        # Woof! - z Animal
rex.fetch()        # Dog-only method
rex.describe()     # dziedziczona z Animal

whiskers.speak()   # Purrrr... Meow! - nadpisana w Cat
whiskers.describe()

# isinstance() sprawdza hierarchię klas
print(isinstance(rex, Dog))     # True
print(isinstance(rex, Animal))  # True - Dog dziedziczy po Animal
print(isinstance(rex, Cat))     # False
````

## 9. 🔹 Metody specjalne (dunder methods)

Metody specjalne (special methods), zwane też dunder methods (od *double
underscore*), to metody o nazwach w postaci `__nazwa__`. Python wywołuje je
automatycznie w odpowiedzi na wbudowane operacje - nie wywołujemy ich
bezpośrednio.

| Metoda | Kiedy wywoływana | Do czego służy |
|---|---|---|
| `__str__(self)` | `print(obj)`, `str(obj)` | czytelna reprezentacja dla użytkownika |
| `__repr__(self)` | `repr(obj)`, konsola | reprezentacja debuggerska, jednoznaczna |
| `__len__(self)` | `len(obj)` | długość / rozmiar obiektu |
| `__init__(self)` | `Klasa(...)` | konstruktor (już znamy) |

Różnica między `__str__` a `__repr__`:
- `__str__` - dla *użytkownika*: `"Rex (Labrador, 3 lata)"`
- `__repr__` - dla *dewelopera*: `"Dog('Rex', 'Labrador', 3)"`

> 💡 Jeśli klasa nie ma `__str__`, Python używa `__repr__` jako fallback.
> Zawsze definiujemy przynajmniej `__repr__` - ułatwia debugowanie.

````python
class Dog:
    def __init__(self, name, breed, age):
        self.name  = name
        self.breed = breed
        self.age   = age

    def __str__(self):
        # czytelna wersja dla użytkownika
        return f"{self.name} ({self.breed}, {self.age} lat)"

    def __repr__(self):
        # wersja dla dewelopera - powinna pozwolić odtworzyć obiekt
        return f"Dog({self.name!r}, {self.breed!r}, {self.age})"


rex = Dog("Rex", "Labrador", 3)

print(rex)        # wywołuje __str__:  Rex (Labrador, 3 lat)
print(repr(rex))  # wywołuje __repr__: Dog('Rex', 'Labrador', 3)

# W liście Python używa __repr__ dla elementów
dogs = [Dog("Rex", "Labrador", 3), Dog("Burek", "Husky", 5)]
print(dogs)


class Basket:
    def __init__(self, items):
        self.items = items

    def __len__(self):
        return len(self.items)

    def __str__(self):
        return f"Basket({self.items})"


b = Basket(["apple", "banana", "cherry"])
print(len(b))   # wywołuje __len__: 3
print(b)        # wywołuje __str__: Basket([...])
````
