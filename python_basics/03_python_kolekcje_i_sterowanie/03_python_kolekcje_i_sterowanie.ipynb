## 1. 📋 Lista (list)

Lista (list) to uporządkowany, **mutowalny (mutable)** zbiór elementów.
Może zawierać elementy różnych typów. Indeksowanie i slicing działają
tak samo jak dla napisów.

```python
my_list = [element1, element2, element3]
```

````python
# Lista może zawierać mieszane typy
items = [1, 'hello', 3.14, True]
print(items)         # [1, 'hello', 3.14, True]
print(type(items))   # <class 'list'>

# Slicing - tak samo jak dla napisów
numbers = [10, 20, 30, 40, 50]
print(numbers[1:4])  # [20, 30, 40]
print(numbers[:3])   # [10, 20, 30]
print(numbers[::2])  # [10, 30, 50]
print(numbers[::-1]) # [50, 40, 30, 20, 10]
````

### Modyfikacja list

Lista jest mutowalna - możemy zmieniać, dodawać i usuwać elementy:

| Operacja | Kod |
|---|---|
| Zmień element | `lst[i] = value` |
| Dodaj na końcu | `lst.append(x)` |
| Wstaw na pozycji `i` | `lst.insert(i, x)` |
| Usuń pierwsze `x` | `lst.remove(x)` |
| Usuń i zwróć element `i` | `lst.pop(i)` |

````python
fruits = ['apple', 'banana', 'cherry']

# Zmiana elementu
fruits[1] = 'blueberry'
print(fruits)   # ['apple', 'blueberry', 'cherry']

# Dodawanie
fruits.append('date')
fruits.insert(1, 'avocado')
print(fruits)   # ['apple', 'avocado', 'blueberry', 'cherry', 'date']

# Usuwanie
fruits.remove('cherry')
popped = fruits.pop(0)   # usuwa i zwraca 'apple'
print(popped)            # apple
print(fruits)            # ['avocado', 'blueberry', 'date']
````

### Sortowanie i inne funkcje

| Funkcja / Metoda | Co robi |
|---|---|
| `len(lst)` | Liczba elementów |
| `lst.count(x)` | Liczba wystąpień `x` |
| `lst.index(x)` | Indeks pierwszego `x` |
| `lst.sort()` | Sortuje listę **w miejscu** (modyfikuje oryginał) |
| `sorted(lst)` | Zwraca **nową** posortowaną listę |
| `sorted(lst, reverse=True)` | Nowa lista posortowana malejąco |

> 💡 `.sort()` zmienia oryginał. `sorted()` tworzy nową listę i nie
> zmienia oryginału. Używamy `sorted()` gdy chcemy zachować oryginał.

````python
numbers = [3, 1, 4, 1, 5, 9, 2, 6, 1]

print(len(numbers))       # 9
print(numbers.count(1))   # 3
print(numbers.index(5))   # 4

# sorted() - nowa lista, oryginał niezmieniony
ascending = sorted(numbers)
print(ascending)   # [1, 1, 1, 2, 3, 4, 5, 6, 9]
print(numbers)     # [3, 1, 4, 1, 5, 9, 2, 6, 1] - niezmieniony!

# .sort() - modyfikuje oryginał
numbers.sort()
print(numbers)     # [1, 1, 1, 2, 3, 4, 5, 6, 9] - zmieniony
````

## 2. 🔒 Krotka (tuple)

Krotka (tuple) to uporządkowana, **niemutowalna (immutable)** kolekcja
elementów. Podobna do listy, ale po utworzeniu nie można jej zmieniać.

Kiedy używamy krotki zamiast listy:
- Dane nie powinny się zmieniać (np. współrzędne GPS, kolor RGB)
- Jako klucze słowników (listy nie mogą być kluczami)

> 💡 Slicing dla krotki działa tak samo jak dla listy - zwraca nową krotkę.

````python
my_tuple = (10, 20, 30, 40, 50)
print(my_tuple)
print(type(my_tuple))   # <class 'tuple'>

# Slicing - tak jak dla list
print(my_tuple[1:4])   # (20, 30, 40)
print(my_tuple[::-1])  # (50, 40, 30, 20, 10)

# Niemutowalność - poniższe wywoła TypeError:
# my_tuple[0] = 99   # TypeError: 'tuple' object does not support item assignment
````

## 3. 📖 Słownik (dictionary)

Słownik (dictionary) to zbiór par **klucz-wartość (key-value pairs)**.
Każdy klucz jest unikalny. Dostęp do wartości odbywa się przez klucz.

```python
my_dict = {key1: value1, key2: value2}
```

| Operacja | Kod |
|---|---|
| Dostęp | `d[key]` |
| Dodaj / zmień | `d[key] = value` |
| Usuń | `del d[key]` |
| Sprawdź klucz | `key in d` |
| Wszystkie klucze | `d.keys()` |
| Wszystkie wartości | `d.values()` |
| Pary klucz-wartość | `d.items()` |

````python
scores = {'Alice': 92, 'Bob': 78, 'Charlie': 85}
print(scores)

# Dostęp
print(scores['Alice'])   # 92

# Dodawanie i zmiana wartości
scores['Diana'] = 95
scores['Bob'] = 80
print(scores)

# Usuwanie
del scores['Charlie']
print(scores)

# Sprawdzenie klucza
print('Alice' in scores)   # True
print('Eve' in scores)     # False
````

````python
phone_book = {'Alice': 500123, 'Bob': 343455, 'Charlie': 612789}

print(list(phone_book.keys()))    # ['Alice', 'Bob', 'Charlie']
print(list(phone_book.values()))  # [500123, 343455, 612789]

for name, number in phone_book.items():
    print(f'{name}: {number}')
````

## 4. 🎯 Zbiór (set)

Zbiór (set) to nieuporządkowana kolekcja **unikalnych** elementów -
duplikaty są automatycznie usuwane. Zbiory nie mają kolejności ani
indeksowania.

> 💡 Szybki sposób usunięcia duplikatów z listy:
> `unique = set([1, 2, 2, 3, 3, 3])` → `{1, 2, 3}`

````python
# Duplikaty są automatycznie usuwane
my_set = {1, 2, 3, 3, 3, 4}
print(my_set)          # {1, 2, 3, 4}
print(type(my_set))    # <class 'set'>

# Usuwanie duplikatów z listy
numbers = [1, 2, 2, 3, 3, 3, 4]
unique = set(numbers)
print(unique)          # {1, 2, 3, 4}

# Sprawdzenie przynależności
print(2 in my_set)     # True
print(5 in my_set)     # False
````

### Operacje na zbiorach (set operations)

Zbiory obsługują matematyczne operacje teoriomnogościowe:

- `a | b` lub `a.union(b)` - suma (wszystkie elementy z obu zbiorów)
- `a & b` lub `a.intersection(b)` - część wspólna (elementy w obu)
- `a - b` lub `a.difference(b)` - różnica (w `a`, ale nie w `b`)
- `a ^ b` lub `a.symmetric_difference(b)` - różnica symetryczna

````python
set_a = {1, 2, 3, 4, 5}
set_b = {3, 4, 5, 6, 7}

print(set_a | set_b)   # {1, 2, 3, 4, 5, 6, 7} - suma
print(set_a & set_b)   # {3, 4, 5}             - część wspólna
print(set_a - set_b)   # {1, 2}                - różnica
print(set_a ^ set_b)   # {1, 2, 6, 7}          - różnica symetryczna
````

## 5. 🔀 Instrukcje warunkowe

Instrukcja warunkowa `if` pozwala wykonać różne bloki kodu w zależności od
spełnienia warunku (condition). Python sprawdza warunki od góry i wykonuje
pierwszy spełniony blok:

```python
if condition:
    code_when_true
elif other_condition:
    code_when_other
else:
    code_otherwise
```

> 💡 Python używa **wcięć (indentation)** zamiast nawiasów klamrowych `{}`.
> Standard to 4 spacje na poziom wcięcia (PEP 8).

````python
score = 73

if score >= 90:
    grade = 'A'
elif score >= 75:
    grade = 'B'
elif score >= 60:
    grade = 'C'
else:
    grade = 'F'

print(f'Grade: {grade}')   # Grade: C
````

### Wyrażenie warunkowe (ternary expression)

Skrócona forma `if`/`else` w jednej linii - przydatna gdy przypisujemy jedną
z dwóch wartości:

```python
value = value_if_true if condition else value_if_false
```

| Podejście | Kod |
|---|---|
| Standardowe if/else | 3 linie: `if ...: result = ...` / `else: result = ...` |
| Wyrażenie warunkowe | 1 linia: `result = ... if ... else ...` |

````python
score = 65

# Standardowe if/else
if score >= 60:
    result = 'passed'
else:
    result = 'failed'
print(result)

# To samo jako wyrażenie warunkowe (ternary)
result = 'passed' if score >= 60 else 'failed'
print(result)
````

### Wcięcia (indentation) - Python vs inne języki

Python wyróżnia się na tle innych języków: używa **wcięć** do oznaczania
bloków kodu zamiast nawiasów klamrowych:

| Język | Blok kodu oznaczany przez |
|---|---|
| C, Java, JavaScript | nawiasy klamrowe `{ }` |
| Pascal | słowa kluczowe `begin` ... `end` |
| Python | wcięcia (indentation) - 4 spacje |

Zaleta: kod jest czytelniejszy i wymusza dobry styl.

> ⚠️ Mieszanie spacji i tabulatorów powoduje błąd `IndentationError`.
> Standardem w Pythonie są 4 spacje na poziom wcięcia.

## 6. 🔄 Pętle (loops)

Pętla (loop) pozwala wielokrotnie wykonać blok kodu - dla każdego elementu
sekwencji lub dopóki spełniony jest warunek. Pętla `for` iteruje po
obiektach iterowalnych (iterable): listach, napisach, zakresach, krotkach.

````python
# Iteracja po napisie
for char in 'Python':
    print(char, end=' ')   # P y t h o n
print()

# Iteracja po krotce
vegetables = ('carrot', 'cauliflower', 'cabbage')
for vegetable in vegetables:
    print(vegetable)

# Zagnieżdżone pętle (nested loops)
for i in [1, 2, 3]:
    for j in ['a', 'b']:
        print(i, j, end='  ')
    print()
````

### Funkcja `range()` - zakres liczb

`range()` generuje sekwencję liczb całkowitych. Używamy jej z pętlą `for`
gdy potrzebujemy iterować określoną liczbę razy:

| Wywołanie | Generuje |
|---|---|
| `range(stop)` | 0, 1, ..., stop-1 |
| `range(start, stop)` | start, start+1, ..., stop-1 |
| `range(start, stop, step)` | co `step` od start do stop-1 |

> 💡 `range()` **nie zawiera** wartości `stop` - `range(5)` generuje
> 0, 1, 2, 3, 4 (pięć elementów, ale bez 5).

````python
# range(stop) - od 0 do stop-1
for i in range(5):
    print(i, end=' ')   # 0 1 2 3 4
print()

# range(start, stop)
for i in range(2, 8):
    print(i, end=' ')   # 2 3 4 5 6 7
print()

# range(start, stop, step) - co 2
for i in range(0, 11, 2):
    print(i, end=' ')   # 0 2 4 6 8 10
print()
````

### Funkcja `enumerate()` - indeks i wartość

Iterując po liście często potrzebujemy zarówno indeksu elementu, jak i
jego wartości. Zamiast ręcznie prowadzić licznik, używamy `enumerate()`:

```python
for i, value in enumerate(iterable):
    ...
```

`enumerate()` zwraca pary `(indeks, wartość)` - rozpakowujemy je przez
tuple unpacking. Opcjonalny parametr `start` pozwala zacząć od innej
liczby niż 0.

> 💡 `enumerate()` jest czytelniejsze i bezpieczniejsze niż ręczny
> licznik `i += 1`. Stosujemy zawsze gdy potrzebujemy indeksu w pętli.

````python
fruits = ['apple', 'banana', 'cherry']

# Bez enumerate - ręczny licznik
i = 0
for fruit in fruits:
    print(i, fruit)
    i += 1

print()

# Z enumerate - elegancko
for i, fruit in enumerate(fruits):
    print(i, fruit)

print()

# enumerate z innym punktem startowym
for i, fruit in enumerate(fruits, start=1):
    print(f"{i}. {fruit}")
````

### Pętla `while` - dopóki warunek spełniony

Pętla `while` wykonuje blok kodu dopóki warunek jest `True`. Używamy jej
gdy nie znamy z góry liczby iteracji.

```python
while condition:
    # kod wykonywany dopóki condition jest True
```

> ⚠️ Zawsze zadbaj o to, żeby warunek `while` w końcu stał się `False`.
> Pętla nieskończona (infinite loop) blokuje program.

````python
# odliczanie
count = 5
while count > 0:
    print(count, end=' ')
    count -= 1
print('Go!')   # 5 4 3 2 1 Go!

# zbieranie kwadratów do listy
squares = []
n = 1
while n <= 5:
    squares.append(n ** 2)
    n += 1
print(squares)   # [1, 4, 9, 16, 25]
````

### `break`, `continue`, `pass`

````python
# break - zatrzymuje pętlę przy n == 5
for n in range(10):
    if n == 5:
        break
    print(n, end=' ')   # 0 1 2 3 4
print()

# continue - pomija parzyste, drukuje nieparzyste
for n in range(1, 10):
    if n % 2 == 0:
        continue
    print(n, end=' ')   # 1 3 5 7 9
print()

# pass - pusty blok (np. do uzupełnienia później)
for n in range(3):
    pass   # TODO: dodać logikę
print('Pętla wykonana')
````
