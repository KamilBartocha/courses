## 4. 🧮 Algorytm

Zanim napiszemy jakikolwiek program, musimy wiedzieć co ten program
ma robić - krok po kroku. Właśnie do tego służy algorytm (algorithm).

Algorytm to skończony ciąg jasno zdefiniowanych kroków prowadzący
do rozwiązania konkretnego problemu. Każdy program komputerowy jest
implementacją jakiegoś algorytmu.

Cechy dobrego algorytmu:

| Cecha | Opis |
|---|---|
| Skończoność | kończy się po skończonej liczbie kroków |
| Jednoznaczność | każdy krok ma dokładnie jedno znaczenie |
| Poprawność | dla poprawnych danych daje poprawny wynik |
| Ogólność | rozwiązuje klasę problemów, nie jeden przypadek |

Pseudokod (pseudocode) to nieformalny sposób zapisu algorytmu.
Używa struktury zbliżonej do kodu, ale bez ścisłych reguł
składniowych - skupiamy się na logice, nie na szczegółach języka.

Przykład pseudokodu - weryfikacja karty kredytowej:

```
JEŻELI numer karty jest ważny TO
    wykonaj transakcję
W PRZECIWNYM RAZIE
    wyświetl komunikat o błędzie
KONIEC WARUNKU
```

> 💡 Pseudokod jest niezależny od języka - ten sam algorytm
> możemy potem zapisać w Pythonie, Javie lub C++.

````python
# Ten sam algorytm - weryfikacja karty - w Pythonie
card_number = "1234-5678-9012-3456"
is_valid = len(card_number) == 19  # uproszczona weryfikacja

if is_valid:
    print("Transakcja zatwierdzona")
else:
    print("Błędny numer karty")
````

## 5. 📊 Schemat blokowy

Przed napisaniem złożonego algorytmu warto go najpierw
zwizualizować - łatwiej wtedy wykryć błędy logiczne. Do tego
służy schemat blokowy (flowchart): graficzna reprezentacja
algorytmu, w której figury geometryczne zastępują słowa.

Podstawowe elementy schematu blokowego:

**Owal (terminal)** - oznacza START lub STOP algorytmu:
```
    ╭─────────╮
    │  START  │
    ╰─────────╯
```

**Prostokąt (process)** - operacja, obliczenie, przypisanie:
```
    ╭──────────────────╮
    │  result = a + b  │
    ╰──────────────────╯
```

**Romb (decision)** - warunek z dwiema gałęziami: TAK lub NIE:
```
         ╭─────────╮
   TAK ◄─┤  x > 0? ├─► NIE
         ╰─────────╯
```

**Strzałka (arrow)** - kierunek przepływu między elementami:
```
    │
    ▼
```

> 💡 W tekstowych schematach romb zastępujemy prostokątem
> z bocznymi strzałkami - ważna jest logika, nie kształt.

Przykład: algorytm sprawdzający parzystość liczby

```
        ╭───────────╮
        │   START   │   ← owal (terminal)
        ╰─────┬─────╯
              │
              ▼
     ╭────────────────╮
     │  number = 7    │   ← prostokąt (process)
     ╰────────┬───────╯
              │
              ▼
     ╭──────────────────╮
     │  r = number % 2  │   ← prostokąt (process)
     ╰────────┬─────────╯
              │
              ▼
         ╭─────────╮             ╭─────────────────────╮
   TAK ◄─┤ r == 0? ├──NIE───────►│ print("nieparzysta")│   ← romb (decision)
         ╰────┬────╯             ╰──────────┬──────────╯
              │                             │
              ▼                             │
     ╭──────────────────╮                   │
     │ print("parzysta")│◄──────────────────┘   ← prostokąt (process)
     ╰────────┬─────────╯
              │
              ▼
        ╭───────────╮
        │   STOP    │   ← owal (terminal)
        ╰───────────╯
```

## 7. 📦 Zmienne

Zmienna to nazwa, pod którą przechowujemy wartość w pamięci
komputera. Tworzymy ją przez przypisanie - znak `=` oznacza:
"niech nazwa X wskazuje na wartość Y".

Co się dzieje w pamięci, gdy piszemy `x = "Alice"`?

```
  krok 1: Python rezerwuje miejsce w pamięci i zapisuje "Alice"
  krok 2: nazwa x zaczyna wskazywać na to miejsce

  x  ──►  [ "Alice" ]
```

Gdy napiszemy `x = "Bob"`, Python:
1. tworzy nowe miejsce w pamięci i zapisuje tam `"Bob"`
```
  x  ──►  [ "Alice" ]
  [ "Bob" ]
```
2. przestawia nazwę `x` na to nowe miejsce
```
  (brak nazwy)  ──►  [ "Alice" ]
  x ──► [ "Bob" ]
```
3. automatycznie usuwa stare `"Alice"` - nie ma już żadnej nazwy,
   która by na nie wskazywała

```
  (brak nazwy) → [ "Alice" ]  ← Python usuwa automatycznie (garbage collector)
  x  ──►  [ "Bob" ]
```

> Zmienna to nie "pudełko z wartością", lecz etykieta
> przyklejona do wartości w pamięci.

> WAŻNE uproszczony sposób przedstawienia tworzenia zmiennych

````python
# inicjalizacja zmiennej
name = "Alice"
print(name)

# zmiana wartości - name wskazuje na nowe miejsce w pamięci
name = "Bob"
print(name)

# dwie zmienne
first_name = "Alice"
last_name = "Smith"
print(first_name, last_name)
````

## 8. ⚙️ Jak działa interpreter?

Gdy uruchamiamy plik `.py`, Python wykonuje go w dwóch etapach.
Zaczyna się od wpisania komendy w terminalu:

```
  ╭─────────────────────────────────╮
  │  $ python hello.py              │  ← Windows
  │  $ python3 hello.py             │  ← macOS / Linux
  ╰─────────────────┬───────────────╯
                    │ uruchamia (przekazując jaki plik)
                    ▼
          ╭─────────────────────╮
          │    python.exe       │   ← Windows:     python.exe
          │                     │     macOS/Linux: python3
          │    (CPython)        │     instalujemy z python.org
          ╰──────────┬──────────╯
                     │ który wewnętrznie robi:
   ──────────────────┼────────────────────────────────────────────
                     │
                     ▼
  ╭──────────────╮  kompilacja  ╭──────────────╮  wykonanie  ╭──────────╮
  │  hello.py    │ ───────────► │  bytecode    │ ──────────► │  Hello!  │
  │              │              │  (.pyc)      │             │          │
  │  print(      │              │              │             │          │
  │   "Hello!")  │              │  LOAD_GLOBAL │             │          │
  │  x = 1 + 2   │              │  LOAD_CONST  │             │          │
  ╰──────────────╯              │  CALL        │             ╰──────────╯
   Kod źródłowy                 ╰──────────────╯               Wynik
   (piszemy my)                    Bytecode                  (Python VM)
                               (tworzy python.exe            wewnątrz
                                automatycznie)               python.exe
```

**`python.exe` / `python3`** to plik wykonywalny instalowany z python.org.
To on łączy wszystkie etapy w całość:

**Etap 1 - Kompilacja do bytecode:**
`python.exe` przetłumaczy nasz kod na uproszczone instrukcje zwane
bytecode (kod bajtowy).

**Etap 2 - Wykonanie przez Python VM:**
`python.exe` uruchamia wbudowaną maszynę wirtualną (Python VM),
która czyta bytecode i wykonuje każdą instrukcję. Standardowa
implementacja Pythona nazywa się **CPython** i jest napisana w C.

| Pojęcie | Opis |
|---|---|
| `python.exe` / `python3` | plik wykonywalny - instalujemy z python.org |
| Kod źródłowy (source code) | plik `.py` - to co piszemy |
| Bytecode | pośredni format binarny; plik `.pyc` w `__pycache__/` |
| Python VM | maszyna wirtualna wykonująca bytecode |
| CPython | standardowa implementacja (`python.exe` to właśnie CPython) |

> 💡 Bytecode jest przenośny między systemami - ten sam `.pyc`
> zadziała na Windows i macOS, o ile wersja Pythona jest taka sama.

> 💡 Moduł `dis` (disassembler) pozwala podejrzeć bytecode
> wygenerowany z naszego kodu - używamy go poniżej.

````python
import dis

def hello():
    print("ALEPH")

dis.dis(hello)


"""
  3           RESUME                   0

  4           LOAD_GLOBAL              1 (print + NULL)
              LOAD_CONST               0 ('ALEPH')
              CALL                     1
              POP_TOP
              LOAD_CONST               1 (None)
              RETURN_VALUE
"""
````

Wynik `dis.dis()` wygląda mniej więcej tak (liczby zależą od wersji Pythona):

```
  linia  offset  instrukcja        argument   (czytelna nazwa)
  ─────  ──────  ────────────────  ─────────  ───────────────────────────
    3            RESUME            0
    4            LOAD_GLOBAL       1          (print + NULL)
                 LOAD_CONST        0          ('Hello!')
                 CALL              1
                 POP_TOP
                 LOAD_CONST        1          (None)
                 RETURN_VALUE
```

Co oznaczają poszczególne kolumny:

| Kolumna | Opis |
|---|---|
| `linia` | numer linii w pliku `.py` (tylko przy pierwszej instrukcji z danej linii) |
| `offset` | pozycja instrukcji w bajtkodzie (0, 2, 4...) |
| `instrukcja` | nazwa operacji (opcode) wykonywanej przez Python VM |
| `argument` | numeryczny parametr instrukcji |
| `(czytelna nazwa)` | co kryje się pod numerem - zmienna, stała, funkcja |

Co oznaczają instrukcje dla `print("Hello!")`:

| Instrukcja | Co robi |
|---|---|
| `RESUME` | wewnętrzna instrukcja startowa - oznacza początek funkcji |
| `LOAD_GLOBAL` | wkłada funkcję `print` na stos (stack) |
| `LOAD_CONST` | wkłada stałą `'Hello!'` na stos |
| `CALL` | wywołuje funkcję ze stosu z 1 argumentem |
| `POP_TOP` | zdejmuje wynik `print()` ze stosu (bo to `None`) |
| `RETURN_VALUE` | zwraca wartość z wierzchołka stosu (niejawne `return None`) |

> 💡 Python VM działa jak kalkulator z **stosem (stack)** - instrukcje
> odkładają wartości na stos i zdejmują je z powrotem. `LOAD_*` odkłada,
> `CALL` i `POP_TOP` zdejmują. Nie musimy tego znać pisząc kod - to
> szczegół implementacyjny, który warto zobaczyć raz.

## 11. 🐍 IDLE - wbudowane środowisko Pythona

IDLE (Integrated DeveLopment Environment) to środowisko programistyczne
dostarczane razem z Pythonem - nie wymaga żadnej dodatkowej instalacji.
To naturalny punkt startowy dla osób zaczynających naukę.

IDLE oferuje dwa tryby pracy:

| Tryb | Opis |
|---|---|
| **Shell** (powłoka) | interaktywne wykonywanie kodu linijka po linijce |
| **Editor** (edytor) | pisanie i uruchamianie całych skryptów `.py` |

Zalety IDLE:

| Cecha | Opis |
|---|---|
| Wbudowany | instaluje się razem z Pythonem, bez konfiguracji |
| Prosty | nie przytłacza - idealny dla nowicjuszy |
| Podświetlanie składni | koloruje słowa kluczowe i literały |
| Interaktywna powłoka | natychmiastowe testowanie wyrażeń |

Ograniczenia IDLE w porównaniu z VS Code czy PyCharm:
- brak zaawansowanego autouzupełniania i analizy typów
- brak systemu wtyczek
- nie nadaje się do zarządzania większymi projektami

> 💡 IDLE jest idealny do pierwszych eksperymentów i nauki składni.
> Przy poważniejszych projektach warto przejść na VS Code lub PyCharm.

## 15. 🔀 Git - kontrola wersji

Git to rozproszony system kontroli wersji (version control
system). Pozwala śledzić zmiany w kodzie, cofać się do
poprzednich wersji i współpracować z innymi programistami.

Podstawowe pojęcia:

| Pojęcie | Opis |
|---|---|
| `repository` | katalog ze śledzonymi plikami |
| `commit` | zapis stanu kodu w danym momencie |
| `branch` | gałąź - izolowana linia rozwoju |
| `merge` | łączenie gałęzi |
| `push` | wysyłanie zmian do zdalnego repo |
| `pull` | pobieranie zmian ze zdalnego repo |

Podstawowe komendy:

| Komenda | Opis |
|---|---|
| `git init` | inicjalizuj nowe repo |
| `git add .` | dodaj pliki do staging area |
| `git commit -m "msg"` | zapisz zmiany z opisem |
| `git status` | sprawdź stan repo |
| `git log --oneline` | historia commitów |
| `git push` | wyślij do GitHub/GitLab |

> 💡 Każde IDE (PyCharm, VS Code) ma wbudowany interfejs
> graficzny dla Gita - nie trzeba pamiętać komend na pamięć.
