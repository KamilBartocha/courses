## 1. 🔹 Czym jest wyjątek?

Wyjątek (exception) to sygnał, który Python wysyła gdy coś pójdzie
nie tak podczas wykonywania programu. Zamiast po prostu przestać
działać bez słowa, interpreter tworzy obiekt wyjątku z opisem błędu
i zatrzymuje wykonywanie bieżącego kodu.

Dzięki wyjątkom możemy:
- dowiedzieć się dokładnie co poszło nie tak,
- obsłużyć błąd w sposób kontrolowany,
- zapobiec awarii całego programu.

Gdy interpreter napotka błąd, wyświetla *traceback* - ścieżkę
prowadzącą do miejsca, w którym wyjątek został zgłoszony.
Poniżej trzy przykłady wbudowanych wyjątków:

> 💡 Czytajmy traceback od dołu - ostatnia linia mówi co poszło
> nie tak, a linie powyżej pokazują przez jakie funkcje
> przeszedł program zanim trafił na błąd.

````python
# ZeroDivisionError - dzielenie przez zero
print(5 * (1 / 0))
````

````python
# NameError - użycie niezdefiniowanej zmiennej
print(2 + x * 3)
````

````python
# TypeError - operacja na niezgodnych typach
print(2 + '2')
````

## 3. 🔹 Obsługa wyjątków

Gdy wyjątek nie jest obsłużony, program się zatrzymuje. Aby temu
zapobiec, używamy bloku `try / except (try/except)`.

| | Bez obsługi | Z obsługą |
|---|---|---|
| Błąd | Program się crashuje | Błąd jest złapany |
| Użytkownik | Widzi traceback | Widzi komunikat |
| Dalszy kod | Nie wykonuje się | Może się wykonać |

### Składnia `try / except`

```python
try:
    # kod, który może zgłosić wyjątek
    ...
except NazwaWyjatku:
    # kod wykonany gdy wyjątek wystąpi
    ...
except InnyWyjątek as err:
    # można przechwycić obiekt wyjątku jako zmienną
    print(f"Error: {err}")
```

Możemy obsłużyć kilka typów wyjątków - każdy w osobnym
bloku `except`.

````python
# Podstawowy try/except
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero!")

# Przechwycenie obiektu wyjątku
try:
    number = int("hello")
except ValueError as err:
    print(f"Error: {err}")

# Kilka bloków except
try:
    items = [1, 2, 3]
    print(items[10])
except IndexError:
    print("Index out of range!")
except TypeError:
    print("Wrong type!")
````

### Blok `else`

Blok `else` wykonuje się tylko wtedy, gdy w bloku `try` **nie**
wystąpił żaden wyjątek. Przydaje się gdy chcemy wyraźnie oddzielić
"szczęśliwy" kod od obsługi błędów.

````python
try:
    result = 10 / 2
except ZeroDivisionError:
    print("Cannot divide by zero!")
else:
    print(f"Result: {result}")  # wykonuje się tylko gdy brak błędu
````

### Blok `finally`

Blok `finally` wykonuje się **zawsze** - niezależnie od tego czy
wyjątek wystąpił, czy nie. Używamy go do "sprzątania": zamykania
plików, połączeń z bazą danych, zwalniania zasobów.

```python
try:
    operacja()
except BłądX:
    obsłuż_błąd()
else:
    następna_operacja()
finally:
    zawsze_wykonaj()  # np. zamknij plik
```

````python
try:
    f = open("06_live_coding.py")
    content = f.read()
except FileNotFoundError:
    print("File not found!")
else:
    print(f"content of file {f.name}:\n{content}")
finally:
    print("Done - always executed")
````

## 4. 🔹 Podnoszenie wyjątków

Możemy sami zgłosić wyjątek używając słowa kluczowego `raise
(raise)`. Przydaje się gdy chcemy wymusić określone warunki w swoim
kodzie - np. walidację danych wejściowych.

```python
raise NazwaWyjatku("komunikat błędu")
```

> 💡 Używajmy konkretnych typów wyjątków (np. `ValueError`,
> `TypeError`) zamiast ogólnego `Exception` - ułatwia to
> debugowanie i łapanie błędów w górę stosu wywołań.

Przykłady użycia `raise`:
- sprawdzenie czy liczba spełnia warunki (np. jest nieujemna),
- sprawdzenie czy argument funkcji jest właściwego typu.

````python
# Przykład 1 - walidacja wartości
x = -1

if x < 0:
    raise ValueError("Value must be non-negative")
````

````python
# Przykład 2 - walidacja typu argumentu
def greet(name):
    if not isinstance(name, str):
        raise TypeError("Name must be a string")
    print(f"Hello, {name}!")

greet(42)
````

## 5. 🔹 Własne wyjątki * (custom exceptions)

Wbudowane wyjątki często nie opisują wystarczająco dobrze problemu
domenowego naszego programu. Możemy tworzyć własne wyjątki przez
dziedziczenie (inheritance) po klasie `Exception` lub jej podklasach.

```python
class NazwaWyjatku(Exception):
    """Opis kiedy ten wyjątek jest zgłaszany."""
    pass
```

Własny wyjątek dziedziczy po istniejącej klasie - dobieramy ją
do rodzaju błędu:

| Klasa bazowa | Kiedy stosować |
|---|---|
| `Exception` | ogólne błędy aplikacji |
| `ValueError` | błędna wartość argumentu |
| `TypeError` | błędny typ argumentu |
| `RuntimeError` | błąd w czasie działania programu |

Zalety własnych wyjątków:
- kod łapiący błąd może reagować inaczej na nasz błąd niż na inne
- nazwa wyjątku sama opisuje co poszło nie tak
- można dodać własne atrybuty z dodatkowymi informacjami

> 💡 Własny wyjątek dziedziczący po `ValueError` może być złapany
> zarówno przez `except NaszWyjątek` jak i `except ValueError`.
> To pozwala na fine-grained lub coarse-grained obsługę błędów.

````python
# Prosta własna klasa wyjątku
class NegativeValueError(ValueError):
    """Zgłaszany gdy wartość jest ujemna."""
    pass


def calculate_discount(price, discount):
    if price < 0:
        raise NegativeValueError(f"Price cannot be negative: {price}")
    if not 0 <= discount <= 100:
        raise ValueError(f"Discount must be 0-100, got: {discount}")
    return price * (1 - discount / 100)


# Łapiemy własny wyjątek
try:
    result = calculate_discount(-10, 20)
except NegativeValueError as e:
    print(f"Custom error: {e}")

# Własny wyjątek jest też ValueError - można go złapać ogólniej
try:
    result = calculate_discount(-10, 20)
except ValueError as e:
    print(f"ValueError (catches custom too): {e}")

# Poprawne wywołanie
print(calculate_discount(100, 20))   # 80.0
````
