## 1. 🔹 Czym sa wzorce projektowe

Wzorzec projektowy (design pattern) to sprawdzone, wielokrotnie stosowane
rozwiazanie powtarzajacego sie problemu projektowego. Wzorce nie sa gotowym
kodem - sa opisem podejscia, ktore mozna zaadaptowac do konkretnej sytuacji.

Historia: w 1994 roku czterech inzynierow - Gamma, Helm, Johnson i Vlissides
(tzw. Gang of Four, GoF) - opublikowalo ksiazke "Design Patterns", opisujac
23 wzorce. Wzorce podzielili na trzy kategorie.

> 💡 Znajomosc wzorcow to wspolny jezyk miedzy programistami. Gdy mowimy
> "uzyj Observer", komunikujemy caly schemat bez tlumaczenia szczegolow.

```python
# Przyklad: bez wzorca - trudno rozszerzyc
class NotificationServiceBad:
    def notify(self, channel: str, message: str) -> None:
        if channel == 'email':
            print(f'Email: {message}')
        elif channel == 'sms':
            print(f'SMS: {message}')
        # dodanie 'push' wymaga modyfikacji tej klasy

# Z wzorcem Strategy - otwarty na rozszerzenie
from abc import ABC, abstractmethod

class Channel(ABC):
    @abstractmethod
    def send(self, message: str) -> None: ...

class EmailChannel(Channel):
    def send(self, message: str) -> None: print(f'Email: {message}')

class SMSChannel(Channel):
    def send(self, message: str) -> None: print(f'SMS: {message}')

class NotificationService:
    def __init__(self, channel: Channel):
        self._channel = channel

    def notify(self, message: str) -> None:
        self._channel.send(message)

NotificationService(EmailChannel()).notify('Hello!')
NotificationService(SMSChannel()).notify('Hi!')
```

## 2. 🔹 Trzy kategorie wzorcow

GoF podzielil 23 wzorce na trzy kategorie:

| Kategoria | Opis | Przyklady |
|---|---|---|
| Kreacyjne (creational) | Tworzenie obiektow | Singleton, Factory, Builder |
| Strukturalne (structural) | Skladanie klas | Adapter, Decorator, Facade |
| Behawioralne (behavioral) | Komunikacja | Observer, Strategy, Command |

Kreacyjne separuja kod tworzacy obiekty od kodu uzywajacego obiektow.
Strukturalne opisuja jak skladac klasy i obiekty w wieksze struktury.
Behawioralne definiuja wzorce komunikacji i przypisywania odpowiedzialnosci.

> 💡 W tym kursie omawiamy po 5-6 wzorcow z kazdej kategorii - te
> najczesciej spotykane w Pythonie.

```python
# Przykladowe wzorce z kazdej kategorii

# KREACYJNY - Singleton: jedna instancja konfiguracji
class Config:
    _instance = None
    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
        return cls._instance

c1, c2 = Config(), Config()
print(f'Singleton: c1 is c2 = {c1 is c2}')  # True

# STRUKTURALNY - Decorator: dodanie zachowania
def uppercase(func):
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs).upper()
    return wrapper

@uppercase
def greet(name: str) -> str:
    return f'hello, {name}'

print(f'Decorator: {greet("alice")}')  # HELLO, ALICE

# BEHAWIORALNY - Strategy: wymienny algorytm sortowania
data = [3, 1, 4, 1, 5, 9, 2, 6]
print(f'Strategy (sorted asc):  {sorted(data)}')
print(f'Strategy (sorted desc): {sorted(data, key=lambda x: -x)}')
```

## 3. 🔹 SOLID: zasady S, O, L

SOLID to piec zasad projektowania obiektowego, opisanych przez
Roberta C. Martina. Pomagaja pisac kod latwy do rozszerzania i testowania.

**S - Single Responsibility Principle (SRP)**
Klasa powinna miec jeden i tylko jeden powod do zmiany.
Jesli klasa zajmuje sie wieloma rzeczami, zmiany w jednej z nich
moga psuc pozostale.

**O - Open/Closed Principle (OCP)**
Oprogramowanie powinno byc otwarte na rozszerzenie, ale zamkniete
na modyfikacje. Nowe funkcjonalnosci dodajemy przez rozszerzenie
(nowe klasy, dziedziczenie, kompozycja), nie przez edytowanie
istniejacego kodu.

**L - Liskov Substitution Principle (LSP)**
Podklasy musza byc podstawialne za klasy bazowe. Jesli funkcja dziala
poprawnie dla obiektu bazowego, musi dzialac tak samo dla kazdego
obiektu pochodnego.

> 💡 SRP jest najczesciej naruszanym SOLID-em. Jesli klasa ma wiecej
> niz jedna metode publiczna dotyczaca roznych odpowiedzialnosci -
> rozwaz podzial.

```python
from abc import ABC, abstractmethod

# SRP: podzial odpowiedzialnosci
class Order:
    def __init__(self, items: list[dict]):
        self.items = items

    def total(self) -> float:
        return sum(i['price'] for i in self.items)

class OrderRepository:           # osobna odpowiedzialnosc: zapis
    def save(self, order: Order) -> None:
        print(f'Saving order: {order.total()}')

class OrderEmailer:              # osobna odpowiedzialnosc: email
    def send_confirmation(self, order: Order) -> None:
        print(f'Email sent for order: {order.total()}')

# OCP: rozszerzanie przez dziedziczenie, nie modyfikacje
class Discount(ABC):
    @abstractmethod
    def apply(self, price: float) -> float: ...

class PercentDiscount(Discount):
    def __init__(self, pct: float): self.pct = pct
    def apply(self, price: float) -> float: return price * (1 - self.pct)

class SeasonalDiscount(Discount):   # nowa klasa, stary kod bez zmian
    def apply(self, price: float) -> float: return price * 0.85

# LSP: Circle i Rectangle sa podstawialne za Shape
class Shape(ABC):
    @abstractmethod
    def area(self) -> float: ...

class Circle(Shape):
    def __init__(self, r: float): self.r = r
    def area(self) -> float: return 3.14159 * self.r ** 2

class Rect(Shape):
    def __init__(self, w: float, h: float): self.w = w; self.h = h
    def area(self) -> float: return self.w * self.h

def total_area(shapes: list[Shape]) -> float:
    return sum(s.area() for s in shapes)

print(total_area([Circle(3), Rect(4, 5)]))
```

```python
# Cwiczenie 2: OCP - Sorter
from abc import ABC, abstractmethod

class Sorter(ABC):
    @abstractmethod
    def sort(self, data: list) -> list: ...

class BubbleSorter(Sorter):
    def sort(self, data: list) -> list:
        ...

class QuickSorter(Sorter):
    def sort(self, data: list) -> list:
        ...

def sort_data(sorter: Sorter, data: list) -> list:  # nie modyfikujemy
    return sorter.sort(data)

numbers = [5, 2, 8, 1, 9]
print(sort_data(BubbleSorter(), numbers))
print(sort_data(QuickSorter(), numbers))
```

```python
# Cwiczenie 3 *(Trudniejsze)*: naruszenie LSP
# hint: Square.height = x powinno tez ustawic width = x
class Rectangle:
    def __init__(self, w: float, h: float):
        self.width = w
        self.height = h
    def area(self) -> float:
        return self.width * self.height

class Square(Rectangle):
    def __init__(self, side: float):
        super().__init__(side, side)
    # Problem: jesli ustawimy tylko width, area bedzie zla

rect = Rectangle(4, 5)
rect.width = 3  # area = 15
square = Square(4)
square.width = 3  # area powinno byc 9, ale jest 12 - LSP naruszone!
print(f'Rectangle area: {rect.area()}')
print(f'Square area (oczekiwane 9, rzeczywiste {square.area()})')
```

## 4. 🔹 SOLID: zasady I, D

**I - Interface Segregation Principle (ISP)**
Klienci nie powinni byc zmuszani do zaleznosci od metod, ktorych
nie uzywaja. Lepiej miec wiele malych, wyspecjalizowanych interfejsow
niz jeden duzy, ogolny.

**D - Dependency Inversion Principle (DIP)**
Moduly wysokopoziomowe nie powinny zalezec od modulow niskopoziomowych.
Oba powinny zalezec od abstrakcji. Abstrakcje nie powinny zalezec od
szczegolow implementacji.

Inaczej: wstrzykuj zaleznosci (dependency injection) przez konstruktor
lub metody, nie tworzac ich wewnatrz klasy.

> 💡 DIP jest kluczem do testowalnosci. Klasa zalezna od abstrakcji
> mozna testowac z mock-iem zamiast prawdziwej implementacji.

> ⚠️ Nie mylic DIP z Dependency Injection (DI) - DI to technika
> implementacji DIP, nie samo zasada.

```python
from abc import ABC, abstractmethod
from typing import Protocol

# ISP: male, wyspecjalizowane protokoly
class Readable(Protocol):
    def read(self) -> str: ...

class Writable(Protocol):
    def write(self, data: str) -> None: ...

class ReadWritable(Readable, Writable): ...  # zlozony protokol

class FileReader:               # implementuje tylko Readable
    def read(self) -> str: return 'file content'

class Logger:                   # implementuje tylko Writable
    def write(self, data: str) -> None: print(f'LOG: {data}')

# DIP: zaleznosc od abstrakcji, nie od konkretnej klasy
class Database(ABC):
    @abstractmethod
    def query(self, sql: str) -> list: ...

class SQLiteDB(Database):
    def query(self, sql: str) -> list:
        return [{'id': 1, 'name': 'Alice'}]

class MockDB(Database):         # latwy do testow
    def query(self, sql: str) -> list:
        return [{'id': 99, 'name': 'TestUser'}]

class UserRepository:           # zalezny od abstrakcji Database
    def __init__(self, db: Database):
        self._db = db           # wstrzyknieta zaleznosc

    def get_users(self) -> list:
        return self._db.query('SELECT * FROM users')

repo_prod = UserRepository(SQLiteDB())
repo_test = UserRepository(MockDB())
print('Production:', repo_prod.get_users())
print('Test:      ', repo_test.get_users())
```

## 5. 🔹 Protokoly Pythona i duck typing

Python uzywa duck typing - jesli obiekt ma odpowiednie metody, mozna
go uzyc zamiast oczekiwanego typu, niezaleznie od hierarchii klas.
"Jesli chodzi jak kaczka i kwacze jak kaczka - to jest kaczka."

`typing.Protocol` (Python 3.8+) pozwala zdefiniowac interfejs strukturalnie:
klasa spełnia protokol jesli ma odpowiednie metody - bez jawnego
dziedziczenia. Nazywamy to structural subtyping.

| | ABC (abc.ABC) | Protocol (typing.Protocol) |
|---|---|---|
| Zaleznosc | jawna (class Foo(MyABC)) | brak (strukturalna) |
| Sprawdzanie | runtime + mypy | glownie mypy |
| Dziedziczenie | wymagane | nie wymagane |
| Uzycie | wewnetrzne API | zewnetrzne / duck typing |

> 💡 Uzywaj `Protocol` gdy chcesz opisac "co obiekt umie", nie
> "z czego dziedziczy". To bardziej pythonowe podejscie.

```python
from typing import Protocol, runtime_checkable

@runtime_checkable
class Drawable(Protocol):
    def draw(self) -> str: ...

class Circle:          # NIE dziedziczy z Drawable
    def draw(self) -> str: return 'O'

class Square:          # NIE dziedziczy z Drawable
    def draw(self) -> str: return '[]'

class TextLabel:       # rowniez nie dziedziczy
    def draw(self) -> str: return 'ABC'

def render_all(shapes: list[Drawable]) -> None:
    for s in shapes:
        print(s.draw())

# Wszystkie trzy klasy spelniaja Drawable bez jawnego dziedziczenia
render_all([Circle(), Square(), TextLabel()])

# @runtime_checkable pozwala uzywac isinstance
print(isinstance(Circle(), Drawable))   # True
print(isinstance('string', Drawable))   # False - brak metody draw
```

3. *(Trudniejsze)* Napisz funkcje `find_max(items)` akceptujaca

```python
# Cwiczenie 1: protokol Comparable i klasa Temperature
from typing import Protocol

class Comparable(Protocol):
    def __lt__(self, other) -> bool: ...
    def __eq__(self, other) -> bool: ...

class Temperature:
    def __init__(self, celsius: float):
        self.celsius = celsius
    def __lt__(self, other: 'Temperature') -> bool: ...
    def __eq__(self, other: 'Temperature') -> bool: ...
    def __repr__(self) -> str: return f'{self.celsius}C'

temps = [Temperature(25), Temperature(10), Temperature(37), Temperature(0)]
print(sorted(temps))
```

```python
# Cwiczenie 2: protokol Summable
from typing import Protocol, runtime_checkable

@runtime_checkable
class Summable(Protocol):
    def __add__(self, other): ...
    def __radd__(self, other): ...

print(isinstance(42, Summable))     # True
print(isinstance(3.14, Summable))   # True
print(isinstance('hello', Summable))  # True (stringi tez maja __add__)
print(isinstance(None, Summable))   # False
```
